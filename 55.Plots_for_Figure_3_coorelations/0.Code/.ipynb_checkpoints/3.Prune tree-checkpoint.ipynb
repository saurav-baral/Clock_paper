```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/50.John's plotting algorithm/"

from Bio import SeqIO

fasta_file = SeqIO.parse(f"{location}/1.Period/Period_selected_few.fas", 'fasta')
species_list = []

for records in fasta_file:
    # if "c-album" in records.id:
    species_name = "_".join(records.id.split("_")[:-1])
    species_list.append(species_name)
    # else:
        # species_name = (records.id.split("-")[0])

    print(species_name)

print(species_list)
```

```python
from Bio import Phylo

tree = Phylo.read(f"{location}/2.Tree/combined_file.fas.treefile_no_branch_length.nwk", "newick")


for tip in tree.get_terminals():
    if tip.name not in species_list:
        tree.prune(tip)

# Optional: Save the pruned tree
Phylo.write(tree, f"{location}/2.Tree/pruned_tree.nwk", "newick")
```
