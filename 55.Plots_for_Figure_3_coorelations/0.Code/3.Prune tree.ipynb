```python
from Bio import Phylo
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/55.Plots_for_Figure_3/1.Gene_trees"
tree = Phylo.read(f"{location}/CWO.nwk", "newick")


for tip in tree.get_terminals():
    if tip.name not in species_list:
        tree.prune(tip)

# Optional: Save the pruned tree
Phylo.write(tree, f"{location}/CWO_pruned.nwk", "newick")
```
