```python
from Bio import Phylo
import drawsvg as draw

tree = Phylo.read("/mnt/g/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/2.Tree/Newick Export_figtree.nwk", 'newick')
species_list = [term.name for term in tree.get_terminals()]
# print(species_list)

with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/18.Common_results/absrel.csv", 'r') as abs_file:
    absrel_results = abs_file.readlines()
results_dic = {}

for i in range(1,len(absrel_results)):
    results_dic [absrel_results[i].split(",")[0]] = absrel_results[i].strip()
    # print(absrel_results[i].split(",")[0])

d = draw.Drawing(9*90, 90*len(species_list)) 

for species_number in range(len(species_list)):
    species = species_list[species_number]
    try:
        results = results_dic[species].split(",")
    except:
        results = [species,"N/A","N/A","N/A","N/A","N/A","N/A","N/A","N/A","N/A"]
    for i in range(1, len(results)):
        # print(results[i])
        if results[i] == "No":
            d.append(draw.Rectangle(10+ 70* i, 20*(species_number+1) + 50*(species_number) + 20, 
                                    50, 30, 
                                    rx=10, ry=10, 
                                    fill='lightgreen', 
                                    stroke='black'))
            
          
           
        if results[i] == "N/A":
            d.append(draw.Rectangle(10+ 70* i, 20*(species_number+1) + 50*(species_number) + 20, 
                                    50, 30, 
                                    rx=10, ry=10, 
                                    fill='none', stroke='black'))
#             
           

        if results[i] == "Yes":
            d.append(draw.Rectangle(10+ 70* i, 20*(species_number+1) + 50*(species_number) + 20, 
                                    50, 30, 
                                    rx=10, ry=10, 
                                    fill='#ff6865', stroke='black'))
            
        d.append(draw.Text(results[i],10, 10+ 70* i + 18, 20*(species_number+1) + 50*(species_number) + 20 + 18, fontWeight='bold'))
        
#         # break
        
#         # if result[i] == N
d.save_svg("/mnt/g/My Drive/Circadian Rhythm Genes Project/18.Common_results/absrel.svg" )
d

```
