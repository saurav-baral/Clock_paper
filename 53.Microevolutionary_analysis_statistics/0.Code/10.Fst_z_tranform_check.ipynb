```python
import numpy as np
import pandas as pd

# Example Fst values
# fst_values = [0.02, 0.05, 0.07, 0.1, 0.35, 0.4, 0.01, 0.08, 0.09]

# df = pd.DataFrame({"FST": fst_values})

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/53.Microevolutionary_analysis_statistics"

fst_data = pd.read_csv(f"{location}/2.Pieris_napi/1.FST_peaks_from_chris/two_pools.skane.lulea.1kb.gt10snps.fst.bed", sep = "\t",low_memory=False, header = None)
fst_data
# print(fst_data)
df = fst_data[3]
df = pd.DataFrame({"FST_raw": fst_data[3].values})
df["FST"] = df["FST_raw"].clip(lower=0)
df = df[df["FST"] >= 0].copy()
print(max(df["FST"].values))
print(min(df["FST"].values))

mu = df["FST"].mean()
sigma = df["FST"].std()

# Z-cutoff
Z_cutoff = 1.645
lower_threshold = mu - Z_cutoff * sigma
upper_threshold = mu + Z_cutoff * sigma

# Mark outliers
df["outlier"] = (df["FST"] < lower_threshold) | (df["FST"] > upper_threshold)

print("Mean FST:", mu)
print("SD FST:", sigma)
print(f"Outlier thresholds: {lower_threshold:.3f} to {upper_threshold:.3f}\n")
print(df)
```
