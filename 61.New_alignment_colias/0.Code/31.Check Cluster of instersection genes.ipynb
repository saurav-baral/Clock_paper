```python
location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/12.Rho_networks/2.Diapausing"

gene_cluster_dict = {}
cluster_count_dict = {}
cluster_count_dict["Full"] = {}
with open(f"{location}/Gene_cluster_degree.csv", 'r') as cluster_file:
    for lines in cluster_file.readlines()[1:]:
        lines_split = lines.strip().split(",")
        gene_cluster_dict[lines_split[0]] = lines_split[2]
        cluster_count_dict["Full"].setdefault(lines_split[2], 0)
        # if lines_split[2] not in cluster_count_dict:
        #     cluster_count_dict[lines_split[2], 0)
        cluster_count_dict["Full"][lines_split[2]] += 1
print(cluster_count_dict)
```

```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/53.Microevolutionary_analysis_statistics"

cluster_species = cluster_count_dict

species_folder_list = ["3.Pararge_aegeria","2.Pieris_napi","1.Lasiomatta_megera"]

for species_folder in species_folder_list:
    cluster_species[species_folder] = {}
    for key in cluster_count_dict["Full"]:
       cluster_species[species_folder][key] = 0 

    with open(f"{location}/{species_folder}/4.Intersections_test/intersection_set.txt",'r') as intersection_file:
        for lines in intersection_file:
            lines_split = lines.strip().split(",")
            cluster = gene_cluster_dict[lines_split[0]]
            # print(lines,cluster)
            cluster_species[species_folder][cluster]+= 1
print(cluster_species)
```

```python
full_Par_Pie_Las = {}
for key in cluster_count_dict["Full"]:
    full_Par_Pie_Las[key] =[]

for key,value in cluster_species.items():
    for cluster,count in value.items():
        full_Par_Pie_Las[cluster].append(str(count))

print(full_Par_Pie_Las)
```
