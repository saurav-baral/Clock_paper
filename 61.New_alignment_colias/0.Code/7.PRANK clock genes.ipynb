```python
from Bio import SeqIO
import os
import subprocess
from concurrent.futures import ProcessPoolExecutor

location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias"
input_dir = f"{location}/7.Merged_sequences/"
output_dir = f"{location}/8.Merged_aligned"

def run_prank(file):
    if file.endswith(".fas"):
        
        input_path = os.path.join(input_dir, file)
        output_path = os.path.join(output_dir, file)
        subprocess.run(f'prank -d="{input_path}" -o="{output_path}" -codon -iterate=10', shell = True)
        # print(f'cd "{location}/3.BUSCO_with_colias_fixed"\nprank -d={file} -o=../3.BUSCO_with_colias_fixed_aligned/{file} -codon -iterate=10', shell = True)
        # assert False
if __name__ == "__main__":
    fasta_files = os.listdir(input_dir)
    print(fasta_files)
    with ProcessPoolExecutor(max_workers=8) as executor:
        executor.map(run_prank, fasta_files)
```
