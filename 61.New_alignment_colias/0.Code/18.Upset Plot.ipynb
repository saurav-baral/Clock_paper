```python
import pandas as pd
from upsetplot import from_indicators, UpSet
import matplotlib.pyplot as plt

# Step 1: Load your data
df = pd.read_csv("/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/8.Upset_plot/counts_5%.csv")  # Replace with your actual file path

# Step 2: Set the 'Gene' column as index (optional but recommended)
df.set_index("Gene", inplace=True)

df = df.astype(bool)


# Step 3: Convert to UpSet-compatible format
upset_data = from_indicators(data=df, indicators=df.columns.tolist())

# Step 4: Create and show the plot
upset = UpSet(upset_data, show_counts=True)
upset.plot()
for text in plt.gcf().findobj(match=plt.Text):
    text.set_fontstyle('italic')

# plt.show()
plt.savefig('/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/8.Upset_plot/upset_output_5%.pdf', format='pdf', bbox_inches='tight')

plt.show()
```
