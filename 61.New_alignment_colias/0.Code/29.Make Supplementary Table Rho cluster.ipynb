```python
location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/12.Rho_networks/2.Diapausing"
import pandas as pd

name_dict = {}

full_table = pd.read_csv(f"{location}/Filtered_set_core_set.csv")
for _, row in full_table.iterrows():
    # print(row)
    if row["Annotation_gene1"] not in name_dict:
        name_dict[row["Annotation_gene1"]] =row["Gene1"] 
    if row["Annotation_gene2"] not in name_dict:
        name_dict[row["Annotation_gene2"]] =row["Gene2"] 
    # assert False
print(len(name_dict))
```

```python
output = "Gene_id,Gene_name,Cluster,Degree"
gene_custer = {}
cluster = pd.read_csv(f"/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/19.Graph tool/gene_clusters_kmeans.csv")
for _, row in cluster.iterrows():
    gene_custer[row["Gene"]] = row["Cluster_kmeans"]

degree = pd.read_csv(f"{location}/Filtered_set_core_set_degree.csv")
for _, row in degree.iterrows():
    output += f'\n{name_dict[row["Gene"]]},{row["Gene"]},{gene_custer[row["Gene"]]},{row["Degree"]}'
print(output)
```

```python
with open(f"{location}/Gene_cluster_degree.csv", 'w') as out_file:
    out_file.write(output)
```
