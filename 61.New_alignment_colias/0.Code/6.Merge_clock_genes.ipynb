```python
location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed"

import os
from Bio import SeqIO

list_of_clock_genes = os.listdir(f"{location}/unTrimmed/Clock")
list_of_new_sequences = os.listdir(f"{location}/3.New_colias/6.Sequences")
# C:\Users\sauba\Desktop\20.RERConverge_trimmed\unTrimmed\Clock
print(list_of_clock_genes)
for files in list_of_clock_genes:
    gene_name = files.split(".")[0]
    output = ''
    fasta_file = SeqIO.parse(f"{location}/unTrimmed/Clock/{files}", 'fasta')
    for records in fasta_file:
        if ("Colias_eurytheme" in records.id) or ("Colias_nastes" in records.id):
            pass
        else:
            output += f">{records.id}\n{records.seq.replace('-','')}\n"
    for new_files in list_of_new_sequences:
        if gene_name in new_files:
            species = new_files.split(".")[1]
            new_fasta_file = SeqIO.parse(f"{location}/3.New_colias/6.Sequences/{new_files}", 'fasta')
            for records_2 in new_fasta_file:
                output += f">{species}\n{records_2.seq[:-3]}\n"
                # print(records_2.id)
    with open(f"{location}/3.New_colias/7.Merged_sequences/{files}", 'w') as out_file:
        out_file.write(output)
                
    
```
