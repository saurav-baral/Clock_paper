```python
import requests
from urllib.parse import quote_plus

def fetch_uniprot_fasta_chunk(ids_chunk):
    """Fetch FASTA sequences from UniProt for a chunk of IDs."""
    query = " OR ".join(f"accession:{uid}" for uid in ids_chunk)
    q_enc = quote_plus(query)
    url = f"https://rest.uniprot.org/uniprotkb/stream?format=fasta&query={q_enc}"
    resp = requests.get(url)
    resp.raise_for_status()
    return resp.text

def fetch_all(ids, chunk_size=3):
    """Fetch all sequences in chunks of given size."""
    all_fasta = []
    for i in range(0, len(ids), chunk_size):
        chunk = ids[i : i + chunk_size]
        fasta = fetch_uniprot_fasta_chunk(chunk)
        all_fasta.append(fasta)
    return "\n".join(all_fasta)

if __name__ == "__main__":
    ids_list = ["A1Z9A8","A1ZAI5","A2BGU9","A3KMH1","A4IG66","A4IGF3","A4IGL7","A5A761","A5LGM7","A5LGM7","A5PJJ7","A6QLU1","A8E657","A8XWX5","A9ULB4","A9ULR1","A9ZSY3","B0FWD3","B0S5Y3","B0VYY5","B0WC46","B0WC46","B2D0J4","B2D0J5","B2D0J5","B2RYW8","B4IMF6","B4JYJ2","B4LS49","B4MQT3","B4QBN3","B5X4H8","C3YWU0","D2GZV9","D2WKD9","F1MJR8","F6QV99","G4WAW9","O01666","O02649","O05389","O05389","O08749","O15160","O16099","O17214","O18598","O18598","O18598","O18640","O31463","O35857","O43614","O43808","O44081","O44252","O44390","O46040","O46108","O57579","O60218","O61231","O61367","O62619","O62742","O70196","O75027","O75897","O76757","O77460","O77784","O88696","O95340","O95433","O95563","O95671","O97463","P00038","P00125","P00129","P00423","P00765","P00768","P00769","P04069","P04069","P05186","P05202","P06603","P07919","P07943","P08397","P08761","P0C7P0","P0CG11","P11029","P11438","P11966","P12261","P12276","P12646","P13255","P15007","P15122","P16066","P16354","P16444","P17336","P17439","P17665","P18101","P18173","P18173","P18173","P18600","P19096","P19178","P19351","P19440","P19891","P20374","P20374","P20374","P20397","P20911","P21914","P22695","P23380","P23919","P24156","P24540","P25183","P25286","P27674","P29786","P30050","P30151","P30736","P31396","P31401","P31402","P31478","P31597","P31937","P32234","P34519","P34528","P34560","P34845","P34847","P34847","P34851","P34853","P35035","P35381","P35502","P35502","P35668","P35720","P35778","P35778","P35778","P35779","P36188","P36362","P37837","P38060","P38942","P40682","P41375","P41565","P41572","P41731","P41777","P42281","P42586","P42765","P42867","P43125","P43307","P43308","P45589","P45589","P46060","P46222","P46430","P46437","P46441","P46562","P46778","P47830","P47947","P48159","P48375","P48506","P48508","P48592","P48604","P48607","P49096","P49914","P50268","P51650","P51659","P51909","P52031","P52899","P53396","P54363","P54622","P54627","P54802","P54960","P54961","P54985","P55096","P55202","P55833","P56389","P56941","P58780","P61023","P61222","P61255","P61265","P62282","P62313","P62752","P62958","P79098","P79345","P79701","P79896","P80147","P80683","P81134","P81399","P81693","P82204","P82251","P82916","P82927","P83088","P83740","P83753","P83851","P84057","P91623","P91660","P91679","P91745","P91745","P91887","P91887","P91929","P91931","P91938","P97355","P97535","P97697","Q01158","Q01989","Q02157","Q02366","Q02375","Q03042","Q03168","Q03755","Q04451","Q05319","Q05825","Q06055","Q06136","Q06HQ7","Q07066","Q07130","Q07152","Q08013","Q08288","Q08623","Q08821","Q09225","Q0E959","Q0IEH7","Q0II59","Q0II70","Q0J4P2","Q0MQB1","Q0MQB3","Q0MQD0","Q0MQE7","Q0Q028","Q0VCU1","Q0ZB77","Q12719","Q13231","Q13336","Q14061","Q148G7","Q16831","Q16GH0","Q16HW7","Q16P20","Q16QL3","Q16W06","Q17058","Q178L7","Q17G75","Q17H91","Q17N71","Q17QK8","Q1ECW2","Q1HPL8","Q1HPS0","Q1HPW4","Q1JPD3","Q1JQ92","Q1JQD2","Q1W674","Q20140","Q20591","Q22067","Q24120","Q24134","Q24154","Q24192","Q24238","Q24251","Q24331","Q24407","Q24439","Q25531","Q26486","Q26486","Q27367","Q27442","Q27591","Q27593","Q28278","Q29397","Q29GB8","Q29IG6","Q29IK4","Q2KIR8","Q2KJI2","Q2L6Z1","Q2T9J0","Q2TA14","Q2TBS2","Q2TJ56","Q2YGT9","Q32PI8","Q3B748","Q3S8M4","Q3SZ71","Q3SZE3","Q3SZW4","Q3T1L5","Q3TLP5","Q3U0L2","Q3UDW8","Q3ULJ0","Q3UVK0","Q3ZBY7","Q3ZCD7","Q49SH1","Q4GX87","Q4GXP2","Q4P9K9","Q4QRH7","Q4R3Y4","Q4R4P4","Q4R4T0","Q4R5E2","Q4V8T0","Q4V9P0","Q53H47","Q53H47","Q54LN4","Q54TR1","Q562C7","Q56A55","Q56FG6","Q575S8","Q57692","Q58820","Q58EB4","Q5BLY5","Q5BLY5","Q5E9T8","Q5EA11","Q5EA79","Q5JTJ3","Q5M8V0","Q5M9G0","Q5NVR2","Q5PQ71","Q5PQR5","Q5R597","Q5R668","Q5R6Z9","Q5R839","Q5RAT4","Q5RBR7","Q5RCA6","Q5RDY1","Q5RE15","Q5REU4","Q5REY3","Q5RF00","Q5RFH0","Q5RFI8","Q5T2R2","Q5T440","Q5T6X4","Q5TM64","Q5TZ51","Q5U2Q3","Q5U2R1","Q5U403","Q5U4T9","Q5XGR8","Q5XIH7","Q5ZJR8","Q5ZK33","Q5ZLN1","Q60611","Q61586","Q62219","Q62452","Q63886","Q642G4","Q642M9","Q64374","Q64442","Q64634","Q66I24","Q68F62","Q69BJ8","Q69CJ9","Q6AYD3","Q6AZW2","Q6DCP1","Q6DG38","Q6DIY9","Q6DRH5","Q6GMR7","Q6GMR7","Q6GQ53","Q6ICL3","Q6MG08","Q6NTT6","Q6NUT3","Q6NVR9","Q6NX12","Q6NZ07","Q6NZN1","Q6P011","Q6P2Z6","Q6P316","Q6PAY8","Q6U841","Q6US81","Q6UX65","Q6WV16","Q6X6Z7","Q6XHI1","Q6XP49","Q6XUZ5","Q6Y306","Q6ZM19","Q70UN9","Q710D6","Q71RS6","Q7JQ07","Q7JUX9","Q7JVE7","Q7K1V5","Q7KML2","Q7KSC4","Q7L1I2","Q7M0E7","Q7M456","Q7M4F3","Q7M4F3","Q7M4F3","Q7M4F3","Q7M4F3","Q7M4F4","Q7M4F4","Q7M4F4","Q7M4F4","Q7PQV7","Q7PWV1","Q7Q4F5","Q7Q6A7","Q7QB45","Q7QBJ0","Q7QC84","Q7RTY1","Q7SXW4","Q7SYV0","Q7T312","Q7T3H9","Q7YR75","Q7YRJ9","Q7Z1B8","Q7ZVF3","Q7ZVZ6","Q7ZYC4","Q80UG8","Q80ZD3","Q868Z9","Q86BJ3","Q86UB9","Q86WA9","Q86WA9","Q8AVF4","Q8BGF9","Q8BMF4","Q8BMF4","Q8BU30","Q8CA03","Q8CGC7","Q8CH36","Q8CHP8","Q8CHP8","Q8CHS4","Q8CIW6","Q8HXG5","Q8HXQ5","Q8I1F4","Q8I1N0","Q8INK9","Q8IUM7","Q8K4D3","Q8L7W2","Q8LFQ6","Q8MMH3","Q8MQS8","Q8MQS8","Q8MSF5","Q8MT06","Q8N5Z0","Q8N8Q8","Q8NFH4","Q8NFV4","Q8NHP6","Q8SPU8","Q8SWU7","Q8SWU7","Q8T0W5","Q8TGM7","Q8TGM7","Q8VCC1","Q8VCE1","Q8VCW8","Q8VEA4","Q8VEJ4","Q8VEM8","Q8VH49","Q8WQI7","Q8WTC1","Q8WY07","Q90512","Q90512","Q91WM3","Q91WQ5","Q91YP3","Q91Z53","Q91ZH7","Q920A6","Q921L7","Q921Y2","Q92536","Q92887","Q93113","Q94225","Q94360","Q94516","Q94518","Q94519","Q94522","Q945L0","Q945U1","Q94920","Q95028","Q95108","Q95327","Q95V39","Q962R6","Q962R9","Q962S7","Q962T5","Q963B6","Q964T2","Q969Z0","Q96RQ3","Q98892","Q99JB2","Q99N89","Q99PL8","Q9BI23","Q9BJZ5","Q9BWD1","Q9CPX8","Q9CQJ8","Q9CQN1","Q9CQX8","Q9CR62","Q9CR98","Q9CW42","Q9CWB7","Q9CXR1","Q9D0B6","Q9D0L4","Q9D1C9","Q9D1L0","Q9D1M4","Q9D289","Q9D3D9","Q9D4D4","Q9D4V0","Q9D8V7","Q9DCW4","Q9DCZ1","Q9EP86","Q9EPR4","Q9EQZ5","Q9GV27","Q9GZU5","Q9H7F0","Q9H7F0","Q9H845","Q9HB07","Q9I7W5","Q9I922","Q9JK41","Q9JLR1","Q9LZG0","Q9MZE0","Q9NB33","Q9NFP1","Q9NHW7","Q9NHW7","Q9NPF2","Q9NRZ7","Q9NVD3","Q9NVH1","Q9NVV4","Q9NVV5","Q9NWS8","Q9NWX6","Q9NYG2","Q9NYV6","Q9P0U1","Q9P2J5","Q9QWZ1","Q9QXA6","Q9QYF1","Q9QYU1","Q9QZW9","Q9R0M5","Q9U4L6","Q9U516","Q9U5N1","Q9U616","Q9U639","Q9U6M0","Q9U8B8","Q9UA35","Q9UGQ3","Q9UHW5","Q9UJ83","Q9UJM8","Q9UK45","Q9UNW1","Q9V3P0","Q9V3W6","Q9V4N3","Q9V4U7","Q9V4U7","Q9V576","Q9V579","Q9V6D6","Q9V6U9","Q9V6Y3","Q9V7D2","Q9V7S5","Q9V7Y2","Q9V9Z1","Q9VA02","Q9VA73","Q9VAI0","Q9VAN0","Q9VAU9","Q9VC49","Q9VCA2","Q9VCA2","Q9VCI0","Q9VCI5","Q9VCR7","Q9VCY8","Q9VD51","Q9VDE5","Q9VE04","Q9VF89","Q9VFJ2","Q9VGN9","Q9VH39","Q9VHN6","Q9VIZ0","Q9VLM5","Q9VLV5","Q9VMU4","Q9VP61","Q9VQ79","Q9VQX4","Q9VQY9","Q9VTN3","Q9VTP4","Q9VU95","Q9VUJ0","Q9VUL9","Q9VUY9","Q9VV43","Q9VVH3","Q9VW26","Q9VWH4","Q9VWP4","Q9VXE0","Q9VY28","Q9VYD7","Q9VZD5","Q9W092","Q9W0H3","Q9W0P5","Q9W107","Q9W141","Q9W2D6","Q9W334","Q9W3T5","Q9W440","Q9W4J5","Q9W4V8","Q9W539","Q9W5N0","Q9W5U2","Q9W5U2","Q9W6I0","Q9WUV0","Q9X248","Q9XSU7","Q9XYN0","Q9XZH6","Q9Y1A3","Q9Y2S2","Q9Y2X3","Q9Y3C4","Q9Y606","Q9Y619","Q9YHT4","Q9YI37","Q9YI37","Q9Z2L6","Q9Z9U1"]

    fasta = fetch_all(ids_list, chunk_size=3)
    print(fasta)

    with open("/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/10.For_blast/Tetrapedia_diversipes_down_diapause_sequences.fasta", "w") as f:
        f.write(fasta)
```

```python
import pandas as pd

# Load BLAST results
cols = ["qseqid", "sseqid", "pident", "length", "qlen", "slen", "evalue", "bitscore"]
df = pd.read_csv("/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/10.For_blast/blast_results_up.tsv", sep="\t", names=cols)

# Compute query coverage
df["qcov"] = df["length"] / df["qlen"] * 100

# Filter: good e-value & coverage
filtered = df[(df["evalue"] < 1e-5) & (df["qcov"] > 50)]

# Sort by query, then best bitscore (or lowest evalue)
filtered = filtered.sort_values(["qseqid", "evalue", "bitscore"], ascending=[True, True, False])

# Pick top hit per query
top_hits = filtered.groupby("qseqid").first().reset_index()

# Save results
top_hits.to_csv("/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/10.For_blast/top_hits_up.tsv", sep="\t", index=False)
print(top_hits)
```
