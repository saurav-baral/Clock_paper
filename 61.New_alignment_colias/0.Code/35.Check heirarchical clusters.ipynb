```python
location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/19.Graph tool"

cluster_dictionary = {}
with open(f"{location}/gene_clusters_by_cocluster.csv", 'r') as cluster_file:
    for lines in cluster_file.readlines()[1:]:
        lines_split = lines.strip().split(",")
        cluster_dictionary.setdefault(lines_split[-1],[])
        cluster_dictionary[lines_split[-1]].append(lines_split[0])
print(cluster_dictionary)
```

```python
from collections import Counter

def is_90_percent_same(lst):
    if not lst:
        return False
    counts = Counter(lst)
    most_common_count = counts.most_common(1)[0][1]
    return most_common_count >= 0.5 * len(lst)

with open(f"{location}/gene_clusters_kmeans.csv", 'r') as cluster_file:
    cluster_lines_list = cluster_file.readlines()[1:]

for cluster, gene_list in cluster_dictionary.items():
    runs_same = 0
    for i in range(100):
        cluster_list = []
        for gene in gene_list:
            for lines in cluster_lines_list:
                if gene in lines:
                    lines_split = lines.split(',')
                    cluster_list.append(lines_split[i+1])
        
        if is_90_percent_same(cluster_list) == 1:
            runs_same += 1
            # print("run ", i+1, cluster_list)
    print(cluster, runs_same)
        # print(gene_list,cluster_list)
# assert False

```
