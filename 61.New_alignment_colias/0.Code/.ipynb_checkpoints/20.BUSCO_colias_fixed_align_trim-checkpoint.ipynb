```python
from Bio import SeqIO
import os
import subprocess
from concurrent.futures import ProcessPoolExecutor

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related"
input_dir = f"{location}/3.BUSCO_with_colias_fixed"
output_dir = f"{location}/3.BUSCO_with_colias_fixed_aligned"

def run_prank(file):
    if file.endswith(".fas"):
        input_path = os.path.join(input_dir, file)
        output_path = os.path.join(output_dir, file)
        # command = f'prank -d={input_path} -o={output_path} -codon -iterate=10'
        print(f'cd "{location}/3.BUSCO_with_colias_fixed"\nprank -d={file} -o=../3.BUSCO_with_colias_fixed_aligned/{file} -codon -iterate=10', shell = True)
        assert False
if __name__ == "__main__":
    fasta_files = os.listdir(input_dir)
    
    with ProcessPoolExecutor(max_workers=8) as executor:
        executor.map(run_prank, fasta_files)
```
