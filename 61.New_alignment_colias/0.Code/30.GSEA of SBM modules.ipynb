```python
location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/61.New_alignment_colias/"

cluster_gene = {}

with open(f"{location}/12.Rho_networks/2.Diapausing/Gene_cluster_degree.csv", 'r') as custer_file:
    for lines in custer_file.readlines()[1:]:
        lines_split = lines.strip().split(",")
        if lines_split[2] not in cluster_gene:
            cluster_gene[lines_split[2]] = []
        cluster_gene[lines_split[2]].append(lines_split[0])
        # assert False
print(cluster_gene)
```

```python
import os
for cluster, gene_list in cluster_gene.items():
    output = "geneid\n"+"\n".join(gene_list)
    try:
        os.mkdir(f"{location}/17.GSEA_Rho-set_clusters/Cluster_{cluster}")
    except:
        pass
    with open(f"{location}/17.GSEA_Rho-set_clusters/Cluster_{cluster}/gene_names.txt", 'w') as out_file:
        out_file.write(output)
    # print(output)
    # assert False
    
```
