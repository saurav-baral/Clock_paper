```python
location = "/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/4.BUSCO_with_colias_fixed_aligned_trimmed"

from Bio import SeqIO
import os

output = "Gene,Count"
list_of_files = os.listdir(f"{location}")
for i,file_name in enumerate(list_of_files):
    # print(i, end = ',')
    fasta_file = SeqIO.parse(f"{location}/{file_name}", 'fasta')
    gene_count = 0
    for records in fasta_file:
        gene_count += 1
    output += f"\n{file_name.split('.')[0]},{gene_count}"
print(output)
```
