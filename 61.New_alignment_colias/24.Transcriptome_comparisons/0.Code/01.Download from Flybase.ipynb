```python
from Bio import Entrez, SeqIO

Entrez.email = "your_email@example.com"

# Search for timeless gene from FlyBase
search = Entrez.esearch(db="nucleotide", term="FBgn0030177")
record = Entrez.read(search)
ids = record["IdList"]

if ids:
    handle = Entrez.efetch(db="nucleotide", id=ids[0], rettype="fasta", retmode="text")
    seq = handle.read()
    print(seq)
    # with open("FBgn0030177.fasta", "w") as f:
    #     f.write(seq)
    print("✅ Sequence saved as FBgn0030177.fasta")
else:
    print("❌ No sequence found for FBgn0030177")
```
