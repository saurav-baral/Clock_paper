```python
def get_transcript(source_species, location):
    list_of_transcripts = os.listdir(f"{location}/1.Annotated Species/{source_species}/Transcripts")
    if "desktop.ini" in list_of_transcripts:
        list_of_transcripts.remove("desktop.ini")


    return(list_of_transcripts[0])
```

```python
source_exon= "Exon7"
target_exon = "Exon6"
offset_1 = 2
offset_2 = 2

source_species = "Danaus_plexippus"
target_species = "Aricia_agestis"


location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/10.Cycle Exon Analysis"
import os
import subprocess
from Bio import SeqIO
list_of_folders_1 = os.listdir(location)
if "Temp" not in list_of_folders_1:
    os.mkdir(f"{location}/Temp")

list_of_folders_2 = os.listdir(f"{location}/Temp")
if "Exon_tester" in list_of_folders_2:
    subprocess.run(f'rm -r "{location}/Temp/Exon_tester"', shell = True)
os.mkdir(f"{location}/Temp/Exon_tester")


query_file = ''

source_exon_file = SeqIO.parse(f"{location}/1.Annotated Species/{source_species}/Transcripts/{get_transcript(source_species, location)}/{source_exon}.fa", "fasta")
for sequences in source_exon_file:
    print(sequences.id, "\n", sequences.seq)
    query_file = f">{sequences.id}\n{sequences.seq[offset_1:].translate()}\n"
    
target_exon_file = SeqIO.parse(f"{location}/1.Annotated Species/{target_species}/Transcripts/{get_transcript(target_species, location)}/{target_exon}.fa", "fasta")
for sequences in target_exon_file:
    print(sequences.id, "\n", sequences.seq)
    query_file += f">{sequences.id}\n{sequences.seq[offset_2:].translate()}\n"

with open(f"{location}/Temp/Exon_tester/query.fa", 'w') as out_file:
    out_file.write(query_file)
    
command = f'"mafft" --localpair --clustalout --maxiterate 16 --reorder "{location}/Temp/Exon_tester/query.fa" > "{location}/Temp/Exon_tester/alignment_clustal_query.txt"'
subprocess.run(command, shell = True, stderr = subprocess.DEVNULL)

with open(f"{location}/Temp/Exon_tester/alignment_clustal_query.txt", 'r') as alignment_file:
    print("".join(alignment_file.readlines()))
```
