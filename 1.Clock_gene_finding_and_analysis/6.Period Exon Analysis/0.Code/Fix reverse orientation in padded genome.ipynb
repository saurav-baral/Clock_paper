```python
from Bio import SeqIO
import os

species = 'Pararge_aegeria'

location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/3.Satyrine/5.For Email_padded"
new_location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/3.Satyrine/6.For Email_fixed"

list_of_species = os.listdir(new_location)
if species not in list_of_species:
    os.mkdir(f"{new_location}/{species}")


fasta_file = SeqIO.parse(f"{location}/{species}/Genome_fragment.fa", 'fasta')

for sequence in fasta_file:
    name = sequence.id
    old_sequence = sequence.seq
    new_sequence = sequence.seq.reverse_complement()
    
with open(f"{new_location}/{species}/Genome_fragment_reversed.fa", 'w') as out_file:
    output = f">{name}\n{new_sequence}"
    out_file.write(output)


```
