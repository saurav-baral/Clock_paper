```python
def Get_genome_name(species, genome_location):
    import os
    list_of_folders = os.listdir(genome_location)
    while True:
        if species in list_of_folders:
            return(species)
        else:
            input("species_missing?")
            break
def Genome_file_name(genome_species_location):
    import os
    list_of_files = os.listdir(genome_species_location)
    for file in list_of_files:
        if file.endswith("_genomic.fna"):
            return(file)

def Get_gff_file_name(genome_species_location):
    import os
    list_of_files = os.listdir(genome_species_location)
    for file in list_of_files:
        if file.endswith("_genomic.gff"):
            return(file)
```

```python
import os
import subprocess

location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/4.Pierinae"
genome_location = "/mnt/f/Genomes_2023-12-26"

list_of_species = os.listdir(f"{location}/1.Blast_result/")
if "desktop.ini" in list_of_species:
    list_of_species.remove("desktop.ini")

print(list_of_species)

for species in list_of_species:

    with open(f"{location}/1.Blast_result/{species}/{species}_coordinates_old.csv",'r') as coordinate_file:
        coordinate_lines = coordinate_file.readlines()


    complement = coordinate_lines[1].split(",")[4]
    if complement == "0":
        start = coordinate_lines[1].split(",")[2]
        stop = coordinate_lines[-1].split(",")[3]
    if complement == "1":
        start = coordinate_lines[-1].split(",")[2]
        stop = coordinate_lines[1].split(",")[3]
    scaffold = coordinate_lines[1].split(",")[1]

    genome_folder_name = Get_genome_name(species, genome_location)

    genome_species_location = f"{genome_location}/{species}"
    genome_file_name = Genome_file_name(genome_species_location)
    
    genome_start = int(start) - 100000
    genome_stop = int(stop) + 100000
    
    list_of_folders = os.listdir(f"{location}/2.TOGA/2.Target_genomes/")
    if species not in list_of_folders:
        os.mkdir(f"{location}/2.TOGA/2.Target_genomes/{species}")
    
    
    subprocess.run(f'samtools faidx "{genome_species_location}/{genome_file_name}" {scaffold}:{genome_start}-{genome_stop} > "{location}/2.TOGA/2.Target_genomes/{species}/target_genome.fa"', shell = True)

    
#     print(species,start,stop,complement)
```
