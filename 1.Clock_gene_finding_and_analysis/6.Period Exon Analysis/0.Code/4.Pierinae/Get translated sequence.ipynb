```python
import os
from Bio import SeqIO
location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/4.Pierinae"

species = "Pieris_brassicae"

list_of_exons = os.listdir(f"{location}/2.Final_output/{species}")
if "desktop.ini" in list_of_exons:
    list_of_exons.remove("desktop.ini")
output = ''
exon_dictionary = {}
max_exon_number = 0
for exon in list_of_exons:
    exon_number = int(exon.split(".")[0].split("_")[1])
    if exon_number > max_exon_number:
        max_exon_number = exon_number

print(exon_number)

for i in range(1,exon_number+1):
    if i != 20:
        exon_name = f"Exon_{i}"
        fasta_file = SeqIO.parse(f"{location}/2.Final_output/{species}/Exon_{i}.fa", 'fasta')
        for records in fasta_file:
    #         print()
#             print(exon_name)
            exon_dictionary[exon_name] = records.seq
# print(output)
# print(output.translate())
# print(list_of_exons)
# print(exon_dictionary)

for i in range(1,exon_number+1):
    if i != 20:
        output += exon_dictionary[f"Exon_{i}"]
#         print(f"Exon_{i}")

print(output, "\n\n\n")
print(output.translate())
```
