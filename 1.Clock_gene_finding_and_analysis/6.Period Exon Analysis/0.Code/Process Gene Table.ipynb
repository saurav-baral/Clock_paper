```python
import re 
import os

location = "H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis"
list_of_species = os.listdir(f"{location}/1.Annotated Species")
if "names" in list_of_species:
    list_of_species.remove("names")
if "desktop.ini" in list_of_species:
    list_of_species.remove("desktop.ini")
species_folder = ''
exon_list = "Exon1,Exon2,Exon3,Exon4,Exon5,Exon6,Exon7,Exon8,Exon9,Exon10,Exon11,Exon12,Exon13,Exon14,Exon15,Exon16,Exon17,Exon18,Exon19,Exon20,Exon21,Exon22,Exon23,Exon24,Exon25,Exon26,Exon27,Exon28,Exon29,Exon30,Exon31,Exon32,Exon33,Exon34".split(",")
for species in list_of_species:
    print(species)
    with open(f"{location}/1.Annotated Species/{species}/Gene_table.txt","r") as gene_table_file:
        gene_table_list = gene_table_file.readlines()
#         print(gene_table_list)

    

    i = 0
    switch = 0
    output = "mRNA Accession\tExon Number\tExon Length\tCoding Length\tGenome Start Point\tGenome Stop Point\n" 
#     output2 = "mRNA Accession\tProtein Accession\tExon Number\tAlignment Score with Bombyx mori\n"
    exon_counter = 0

    while i < (len(gene_table_list)):

        if "Exon table for  mRNA  " in gene_table_list[i]:
            gene_table_line = gene_table_list[i]
            while re.search("[ ]{2,}", gene_table_line):
                gene_table_line = gene_table_line.replace("  "," ")

            mrna_accession = gene_table_line.split(" ")[4]
        
            switch = 1
            i += 3
        if switch == 1:

            gene_table_line = gene_table_list[i]
            while re.search("[\t]{2,}", gene_table_line):
                gene_table_line = gene_table_line.replace("\t\t","\t")
            print(len(gene_table_line.split("\t")))
            if len(gene_table_line.split("\t")) == 7 or len(gene_table_line.split("\t")) == 6:
                exon_length = gene_table_line.split("\t")[4]
                coding_exon_length =  ((gene_table_line.split("\t")[5]))
                genome_start_point,genome_end_point = ((gene_table_line.split("\t")[0]).split("-"))
#                 print((gene_table_line.split("\t")[1]).split("-"))
                output += f"{mrna_accession}\t{exon_list[exon_counter]}\t{exon_length.rstrip()}\t{coding_exon_length.rstrip()}\t{genome_start_point}\t{genome_end_point}\n"
                exon_counter += 1
            elif len(gene_table_line.split("\t")) == 4:
                exon_length = gene_table_line.split("\t")[2]
                genome_start_point,genome_end_point = ((gene_table_line.split("\t")[0]).split("-"))
                output += f"{mrna_accession}\t{exon_list[exon_counter]}\t{exon_length.rstrip()}\t0\t{genome_start_point}\t{genome_end_point}\n"
                exon_counter += 1

            if switch == 1 and re.search("^mRNA",gene_table_list[i]):
                switch = 0
                print(exon_counter)
                output += "\n\n"
                exon_counter = 0

        i += 1

    with open(f"{location}/1.Annotated Species/{species}/Gene_table_noncoding_processed.tsv",'w') as out_file:
        out_file.write(output)
```
