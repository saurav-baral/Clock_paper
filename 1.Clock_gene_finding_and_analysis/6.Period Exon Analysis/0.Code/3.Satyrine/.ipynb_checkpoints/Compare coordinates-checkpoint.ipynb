```python
import os
species_list = ["Coenonympha_glycerion","Elymnias_hypermnestra","Erebia_aethiops","Erebia_ligea","Hipparchia_semele","Lasiommata_megera","Maniola_hyperantus","Maniola_jurtina","Melanargia_galathea","Oeneis_ivallda","Pararge_aegeria"]
species = "Coenonympha_glycerion"
location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/3.Satyrine/1.Blast_result"
list_of_files = os.listdir(f"{location}/{species}")
coordinate_files_list = []
output = f'{species}'
for file in list_of_files:
    if file.endswith("_coordinates.csv"):
        with open(f"{location}/{species}/{file}", 'r') as coordinate_file:
            coordinate_file_list = coordinate_file.readlines()
    elif file.endswith(".csv"):
        # if "_old" in file:
        #     # output += "\told\tQuery_name"
        # else:
        #     species_name = file.split(".")[1]
        #     output += f"\t{species_name}\tQuery_name"
        coordinate_files_list.append(file)

# print(output)
# print(coordinate_files_list)

for i in range(1,len(coordinate_file_list)):
    coordinate_split = (coordinate_file_list[i].split(","))
    # print(coordinate_split)
    exon = coordinate_split[6].split("Exon")[1][1:]
    start_coordinate = coordinate_split[2]
    stop_coordinate = coordinate_split[3]
    output += f"\nExon_{exon}\tStart\tStop"
    for alternative_coordinate_file in coordinate_files_list:
        with open(f"{location}/{species}/{alternative_coordinate_file}", 'r') as alternative_coordinate_file_open:
            alternative_coordinate_file_open_list = alternative_coordinate_file_open.readlines()
        if "_old" in alternative_coordinate_file:
            species_name = "Old"
        else:
            species_name = alternative_coordinate_file.split(".")[1]
        output += f"\n{species_name}"
        match = 0
        for j in range(1, len(alternative_coordinate_file_open_list)):
            alternative_coordinate_file_split = alternative_coordinate_file_open_list[j].split(",")
            alternative_start_coordinate = alternative_coordinate_file_split[2]
            alternative_stop_coordinate = alternative_coordinate_file_split[3]
            if alternative_start_coordinate == start_coordinate:
                output += f"\tMatch"
                match = 1
                if alternative_stop_coordinate == stop_coordinate:
                    output += f"\tMatch"
                else:
                    output += f"\tNot Match"
                break

        if match == 0:
            output += f"\tNot Match\tNot Match"
        # output += "\n"
            # if alternative_start_coordinate == start_coordinate:
print(output)
with open(f"{location}/{species}/match_mismatch.tsv", 'w') as out_file:
    out_file.write(output)
```
