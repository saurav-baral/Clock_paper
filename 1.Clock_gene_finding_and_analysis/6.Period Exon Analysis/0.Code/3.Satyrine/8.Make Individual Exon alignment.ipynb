```python
def check_alignment_folder(output_location):
    list_of_folders = os.listdir(output_location)
    if "3.Test Alignment" not in list_of_folders:
        os.mkdir(f"{output_location}/3.Test Alignment")
```

```python
import os
from Bio import SeqIO
output_location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/3.Satyrine"

check_alignment_folder(output_location)
total_exons = 27
list_of_species = os.listdir(f"{output_location}/2.Final_output")
list_of_species.remove('desktop.ini')
# print(list_of_species)


for i in range(1,total_exons+1):
    exon = f"Exon_{i}"
    output = ''
    for species in list_of_species:
    
        try:
            fasta_file = SeqIO.parse(f"{output_location}/2.Final_output/{species}/{exon}.fa",'fasta')
            for records in fasta_file:
                overhang = int(records.id.split("_")[-3])
                sequence = records.seq[overhang:]
                exon_sequence = f">{records.id}\n{sequence}"
#             with open(f"{output_location}/2.Final_output/{species}/{exon}.fa",'r') as exon_file:
#                 exon_sequence = ''.join(exon_file.readlines())
            output += exon_sequence +"\n"
        except:
            pass
# #         print(exon_sequence)
        
    with open(f"{output_location}/3.Test Alignment/{exon}_aligned.fas",'w') as out_file:
        out_file.write(output)
              
#     assert False
```
