```python
import os
import re
```

```python
list_of_species = os.listdir("H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/1.Annotated Species")
query_species = "5.Bicyclus_anynana"
query_transcript = "XM_024088150.2"
list_of_species.remove(query_species)
if "desktop.ini" in list_of_species:
    list_of_species.remove("desktop.ini")
list_of_genomes = os.listdir("F:/Genomes_2023-12-26")
```

```python
query_location = f"H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/0.Query/{query_species}/{query_transcript}/query_protein.fa"
genome_file = ''
output = ''
for species_folder in list_of_species:
    species = species_folder.split(".")[1]
    blast_output_folders = os.listdir("H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/1.Blast_output")
    if species not in blast_output_folders:
        os.mkdir(f"H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/1.Blast_output/{species}")
    
    
    for genome in list_of_genomes:
        if species == genome:
            files_in_genome_folder = os.listdir(f"F:/Genomes_2023-12-26/{genome}")
            for file in files_in_genome_folder:
                if re.search('_genomic.fna$', file):
                    genome_file = file
                    break
    
    output += f'tblastn -seg no -query "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/0.Query/{query_species}/{query_transcript}/query_protein.fa" -db "/mnt/f/Genomes_2023-12-26/{species}/{genome_file}" -num_alignments 3 -out "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/1.Blast_output/{species}/{species}_blast_out.htm" -html \n\n tblastn -seg no -query "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/0.Query/{query_species}/{query_transcript}/query_protein.fa" -db "/mnt/f/Genomes_2023-12-26/{species}/{genome_file}" -num_alignments 3 -out "/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/1.Blast_output/{species}/{species}_blast_out.txt"\n\n'
    
with open("H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/tblastn.sh", "w", encoding="utf-8") as outfile:
    outfile.write(output)
#             print("here",genome)
#     print(species_folder)

```
