```r
leisr_data = read.table("G:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/03.LEISR/LEISR_rate.txt", sep = "\t", header = T)
```

```r
library(ggplot2)
library(cowplot)
```

```r
exon_plot = ggplot() + geom_line(data = leisr_data, aes(x = Site, y = Rate), color = "#1B7593")+  ylim(0, 5) + theme_cowplot(12) 
exon_plot
```

```r
pdf(file = "G:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/03.LEISR/LEISR_rate.pdf",   # The directory you want to save the file in
    width = 10, # The width of the plot in inches
    height = 4)
exon_plot
```
