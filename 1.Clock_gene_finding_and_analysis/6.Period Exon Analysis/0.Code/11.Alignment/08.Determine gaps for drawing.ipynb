```python
from Bio import SeqIO


location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/"

alignment_file = SeqIO.parse(f"{location}/period.masceDNA.fas", 'fasta')
output = ''
for records in alignment_file:
    output += records.id+ ","
    gap_start = 0
    for i in range(len(records.seq)):
        if records.seq[i] == "-" and gap_start == 0:
            output += f"{i+1}-"
            gap_start = 1
        if gap_start == 1 and records.seq[i] != "-":
            output += f"{i+1},"
            gap_start = 0
    if gap_start == 1:   
        output += f"{i+1}\n"
    else:
        output =output[:-1]+"\n"
    
print(output)
with open(f"{location}/06.For figure/gap_position.txt", 'w') as out_file:
    out_file.write(output)
```
