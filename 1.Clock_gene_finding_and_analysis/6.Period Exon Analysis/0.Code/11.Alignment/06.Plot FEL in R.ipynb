```python
library(ggplot2)
library(cowplot)

FEL_data_codon = read.table("G:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/04.FEL/FEL_rate.txt", sep = "\t", header = T)
FEL_plot = ggplot() + geom_line(data = FEL_data_codon, aes(x = Site, y = Omega), color = "#1B7593") +
  ylim(0,2)+ 
  geom_hline(aes(yintercept=1),color = "salmon")+
  scale_x_continuous(breaks = pretty(FEL_data_codon$Site, n = 30)) +
  theme_cowplot(12) +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) 
FEL_plot
```

```python
pdf(file = "G:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/04.FEL/FEL_rate.pdf",   # The directory you want to save the file in
    width = 10, # The width of the plot in inches
    height = 4)
FEL_plot
```
