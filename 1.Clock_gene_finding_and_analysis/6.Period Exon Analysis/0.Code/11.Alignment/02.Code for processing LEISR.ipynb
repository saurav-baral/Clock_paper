```python
import json 

with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/03.LEISR/period.masceDNA.fas.LEISR.json", 'r') as f:
  data = json.load(f)

# print(data["MLE"])
output = "Site\tRate"
dataset = (data["MLE"]["content"]["0"])
for  i in range(len(dataset)):
    output += f"\n{i+1}\t{dataset[i][0]}"
    # print(dataset[i])

with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/03.LEISR/LEISR_rate.txt", 'w') as out_file:
  out_file.write(output)


output_split  = output.split("\n")
output2 = "Site\tRate"
counter = 1
site_number = 1
while counter < len(output_split):
    new_rate = (float(output_split[counter].split("\t")[1]) + float(output_split[counter+1].split("\t")[1]) + float(output_split[counter+2].split("\t")[1]))/3
    output2 += f"\n{site_number}\t{new_rate}"
    site_number += 1
    counter += 3
with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/03.LEISR/LEISR_rate_codon.txt", 'w') as out_file:
  out_file.write(output2)
# print(output2)
```
