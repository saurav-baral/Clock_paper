```python
from Bio import SeqIO

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment"

output1 = ""
output2 = ''

untranslated = SeqIO.parse(f"{location}/00.Alignment/period_aligned_prank.fas.best.fas", 'fasta')

for records in untranslated:
    output1 += f">{records.id}\n{records.seq.replace('-','')}\n\n"
    output2 += f">{records.id}\n{records.seq.translate()}\n\n"
print(output1)
with open(f"{location}/01.Trimal/period_aligned_prank.fas.best.ungapped.fas", 'w') as out_file:
    out_file.write(output1)

with open(f"{location}/01.Trimal/period_aligned_prank.fas.best.translated.fas", 'w') as out_file:
    out_file.write(output2)
```
