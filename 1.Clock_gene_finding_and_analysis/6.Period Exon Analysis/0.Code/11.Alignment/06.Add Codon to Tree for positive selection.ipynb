```python
from Bio import SeqIO
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment/"
pos = 1276
alignment = SeqIO.parse(f"{location}/period.masceDNA.fas", 'fasta')
with open(f"{location}/period.masceDNA.fas.treefile", 'r') as tree_file:
    tree= tree_file.readlines()[0]

for records in alignment:
    codon = records.seq[(pos*3)-3:pos*3]
    spaces = 40-len(records.id)
    tree = tree.replace(records.id, f"{records.id}{'_'*spaces}{codon}_{codon.translate()}")
    # print(records.seq[(pos*3)-3:pos*3])
    # print(records.seq[(pos*3)-3:pos*3].translate())

with open(f"{location}/04.FEL/period.masceDNA.fas_pos_sele_fel.treefile", 'w') as out_file:
    out_file.write(tree)
```
