```python
import os
from Bio import SeqIO
```

```python
blast_output_location = "H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/1.Blast_output"
annotated_genome_location = "H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/1.Annotated Species"
reblast_location = "H:/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/2.Blast/4.Re_Blast"
species = "Pararge_aegeria"
coordinate_file = f"{species}_coordinates.csv"

genome_location = "F:/Genomes_2023-12-26"
list_of_files_in_genome_folder = os.listdir(f"{genome_location}/{species}")
for file in list_of_files_in_genome_folder:
    if file.endswith("_genomic.fna"):
        genome_file = file
# print(genome_file)
```

```python
with open(f"{blast_output_location}/{species}/{coordinate_file}", 'r') as coordinate_open_file:
    coordinate_file_lines = coordinate_open_file.readlines()
    
exon_first_line, exon_last_line = coordinate_file_lines[1].split(","), coordinate_file_lines[-1].split(",")
if ("Y" in exon_first_line[5]) or ("Y" in exon_last_line[5]):
    print("ERROR!!!")
    print(f"{coordinate_file_lines[0]}\n{coordinate_file_lines[1]}\n{coordinate_file_lines[-1]}")
    assert False
          

complement,scaffold = exon_first_line[4], exon_first_line[1]

if complement == "0":
    gene_start = min(int(exon_first_line[2]),int(exon_first_line[3]))
    gene_end = max(int(exon_last_line[2]),int(exon_last_line[3]))

if complement == "1":
    gene_start = min(int(exon_last_line[2]),int(exon_last_line[3]))
    gene_end = max(int(exon_first_line[2]),int(exon_first_line[3]))

print(gene_start, gene_end)


print(exon_first_line)
print(exon_last_line)
# print(coordinate_file)
```

```python
genome = SeqIO.parse(f"{genome_location}/{species}/{genome_file}", "fasta")
for entries in genome:
    if scaffold == entries.id:
        print(scaffold)
        gene_sequence = entries.seq[gene_start-1:gene_end]
        break
# print(gene_sequence)
```

```python
list_of_annotated_genomes = os.listdir(annotated_genome_location)
annotated_species_name = ''
for annotated_species in list_of_annotated_genomes:
    if annotated_species.endswith(species):
        annotated_species_name = (annotated_species)
        break
if annotated_species_name == '':
    print(f"Error with annotated species name")
    assert False
    
list_of_files_inside_annotated_species_folder = os.listdir(f"{annotated_genome_location}/{annotated_species_name}")

if "Period_gene_genomic_sequence" not in list_of_files_inside_annotated_species_folder:
    os.mkdir(f"{annotated_genome_location}/{annotated_species_name}/Period_gene_genomic_sequence")

with open(f"{annotated_genome_location}/{annotated_species_name}/Period_gene_genomic_sequence/period_gene_genomic_sequence.fa",'w') as out_file:
    output = f">{species}_genome_{scaffold}_{gene_start}_{gene_end}\n{gene_sequence}"
    out_file.write(output)


```
