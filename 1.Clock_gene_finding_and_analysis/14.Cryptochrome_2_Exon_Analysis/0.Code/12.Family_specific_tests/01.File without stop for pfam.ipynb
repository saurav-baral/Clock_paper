```python
from Bio import SeqIO
output = ''

fasta_file = SeqIO.parse("/mnt/g/My Drive/Circadian Rhythm Genes Project/7.Timeless Exon Analysis/12.Family_specific_tests_timeless/Combined/full_alignment.fas", 'fasta')
for records in fasta_file:
    if "*" in records.seq.translate():
        if "*" not in records.seq.translate()[:-1]:
            output += f">{records.id}\n{records.seq[:-3]}\n"
        else:
            print(records.id, "errror")
with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/7.Timeless Exon Analysis/12.Family_specific_tests_timeless/Combined/full_alignment_nostop.fas", 'w') as out_file:
    out_file.write(output)
    
```
