```python
from Bio import SeqIO
import os
import subprocess

sequence_details = "Luehdorfia_chinensis,JAHVAK010000144.1,3181536,3181618,0"

species,scaffold,start,stop,complement = sequence_details.split(",")

genome_location = "/mnt/f/Genomes_2023-12-26"
frame = 1

subprocess.run(f'cd {genome_location}/{species}\nsamtools faidx {get_genome_file(genome_location,species)} {scaffold}:{start}-{stop} > test.txt  ', shell = True)

fasta_file = SeqIO.parse(f'{genome_location}/{species}/test.txt', 'fasta')

for records in fasta_file:
    if complement == "1":
        sequence = (records.seq.reverse_complement())
    else:
        sequence = (records.seq)

print(sequence)

print("frame1:",sequence.translate())
print("frame2:",sequence[1:].translate())
print("frame3:",sequence[2:].translate())
```

```python
def get_genome_file(genome_location,species):
    list_of_files_in_genome_folder = os.listdir(f"{genome_location}/{species}")
    for file in list_of_files_in_genome_folder:
        if file.endswith("_genomic.fna"):
            genome_file = file
    return(genome_file)  
```
