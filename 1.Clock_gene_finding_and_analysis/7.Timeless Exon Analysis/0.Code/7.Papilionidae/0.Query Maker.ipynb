```python
import os
from Bio import SeqIO
import re
```

```python
family_group = "7.Papilionidae"

trascript_source_location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/7.Timeless Exon Analysis/1.Annotated Species"
source_species = "Papilio_xuthus"
source_transcript = "XM_013325762.1"



query_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/7.Timeless Exon Analysis/{family_group}/1.Query"
try:
    os.mkdir(query_location)
except:
    pass
folders_in_query = os.listdir(query_location)
if source_species not in folders_in_query:
    os.mkdir(f"{query_location}/{source_species}")
transcript_folders = os.listdir(f"{query_location}/{source_species}")

if source_transcript not in transcript_folders:
    os.mkdir(f"{query_location}/{source_species}/{source_transcript}")
```

```python
list_of_exons = os.listdir(f"{trascript_source_location}/{source_species}/Transcripts/{source_transcript}")
# print(list_of_exons)
if "desktop.ini" in list_of_exons:
    list_of_exons.remove("desktop.ini")

output = ''
exon_dictionary = {}

for exon in list_of_exons:
    exon_number = int(exon.split(".")[0].split("Exon")[1])
    
    sequence = SeqIO.parse(f"{trascript_source_location}/{source_species}/Transcripts/{source_transcript}/{exon}", "fasta")

    for record in sequence:
        # print(record.id)
        # print(record.seq)
        # translated_exon = record.seq[left_overhang:].translate()
        exon_dictionary[f"Exon_{exon_number}"] = record.seq
exon_overhang = {}
exon_structure = ""
for exon in range(1,len(exon_dictionary)+1):
    if exon == 1:
        left_overhang = 0
        right_overhang = len(exon_dictionary[f"Exon_{exon}"]) % 3
        exon_overhang[f"Exon_{exon}"] = [left_overhang,right_overhang]
    else:
        left_overhang = (3-(exon_overhang[f"Exon_{exon-1}"][1]))%3
        right_overhang = (len(exon_dictionary[f"Exon_{exon}"])-left_overhang) % 3
        exon_overhang[f"Exon_{exon}"] = [left_overhang,right_overhang]
    
    exon_length = len(exon_dictionary[f'Exon_{exon}'])
    exon_structure += f"Exon {exon} : {exon_length} : {exon_length/3} : {left_overhang} + {exon_length-left_overhang-right_overhang} * 3 + {right_overhang}\n"
    translated_sequence = exon_dictionary[f'Exon_{exon}'][left_overhang:].translate()
    if "*" in translated_sequence:
        print(f"\n\nExon_{exon}\nstop_codon_here\n\n")
    # print(f"Exon_{exon}\n{translated_sequence}")
    with open(f"{query_location}/{source_species}/{source_transcript}/query_Exon_{exon}.fa", 'w') as query_file:
        output1 = f">{source_species}_{source_transcript}_Frame_{left_overhang}_rightoh_{right_overhang}_query_Exon_{exon}\n{translated_sequence}"
        query_file.write(output1)
    output += output1+"\n\n"

with open(f"{query_location}/{source_species}/{source_transcript}/query_protein.fa", 'w') as query_file:
    query_file.write(output)
    
    # print(f"Exon_{exon}",exon_dictionary[f"Exon_{exon}"])

# for key, value in exon_overhang.items():
#     print(key,value)
print(exon_structure)
```
