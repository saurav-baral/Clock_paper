```python
import os
from Bio import SeqIO

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis"
family_group_list = ["3.Satyrine","4.Pierinae","5.Coliadinae","6.Leptidea","8.Heliconiinae_Danainae","9.Papilionidae","10.Lycaenidae"]
gene = "period"
output = ''
output_2 = 'Species\tExons'

for family_group in family_group_list:
    list_of_species = os.listdir(f"{location}/{family_group}/2.Final_output/")
    
    if "desktop.ini" in list_of_species:
        list_of_species.remove("desktop.ini")
    
    
    for species in list_of_species:
        print(species)
        output_2 += f"\n{species}"
        individual_species_sequence = ''
        list_of_exon_files = os.listdir(f"{location}/{family_group}/2.Final_output/{species}")
        for i in range(1, 30):
            try:
                exon_file = SeqIO.parse(f"{location}/{family_group}/2.Final_output/{species}/Exon_{i}.fa", 'fasta')
                output_2 +=f"\tExon_{i}" 
            except:
                output_2 +=f"\t" 
                continue
            for records in exon_file:
                # print(records.id,records.seq)
                individual_species_sequence += records.seq 
        # print(list_of_exon_files)
        if "*" in individual_species_sequence.translate()[:-1]:
            print(individual_species_sequence.translate())
            assert False
        output += f">{species}-{gene}\n{individual_species_sequence}\n\n"
        
        
    # print(output_2)
    # assert False
with open(f"{location}/11.Alignment/{gene}.fas", 'w') as out_file:
    out_file.write(output)

with open(f"{location}/11.Alignment/exon_details.txt", 'w') as out_file_2:
    out_file_2.write(output_2)
```
