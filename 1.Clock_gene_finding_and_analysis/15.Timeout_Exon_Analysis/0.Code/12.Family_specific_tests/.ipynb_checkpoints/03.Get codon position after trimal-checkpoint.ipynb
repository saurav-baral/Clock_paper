```python
import json
from Bio import SeqIO


with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/12.Family_specific_tests/Pieridae/Pieridae_period_nostop.fas.aligned.fas.trimal.fas.BUSTED.json", 'r') as f:
    busted_result_trimmed = json.load(f)

# print(busted_result_trimmed["Evidence Ratios"]["constrained"][0])
evidence_ratio_list = busted_result_trimmed["Evidence Ratios"]["constrained"][0]
sites_with_positive_selection = []
for site in range(len(evidence_ratio_list)):
    if float(evidence_ratio_list[site]) > 10:
        sites_with_positive_selection.append([site+1, evidence_ratio_list[site]])
print(sites_with_positive_selection)
with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/12.Family_specific_tests/Pieridae/1.trimal_outputs/columns_kept.txt",'r') as column_file:
    column_list = column_file.readlines()
sites = []
for site_number in  sites_with_positive_selection:
    # print(site_number[0])
    print(column_list[0].strip().split("\t")[1].split(", ")[site_number[0]], site_number[1])
    sites.append([column_list[0].strip().split("\t")[1].split(", ")[site_number[0]], site_number[1]])    

aligned_file = SeqIO.parse("/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/12.Family_specific_tests/Pieridae/Pieridae_period_nostop.fas.aligned.fas.best.fas", 'fasta')

for records in aligned_file:
    print(records.id)
    for site in sites:
        
        print(len(records.seq.translate()[:int(site[0])-1].replace("-",'')), site[1])
    
```
