```python
from Bio import SeqIO
import os
import subprocess
family, gene = "Lycaenidae", "period"
location = f"/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/12.Family_specific_tests/{family}"
file_name = f'{family}_{gene}.fas'

subprocess.run(f'cd {location}\nprank -d={file_name} -o={file_name}.aligned.fas -codon', shell = True, stderr = subprocess.DEVNULL)



# location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/11.Alignment"

output1 = ""
output2 = ''

untranslated = SeqIO.parse(f"{location}/{file_name}.aligned.fas.best.fas", 'fasta')

for records in untranslated:
    output1 += f">{records.id}\n{records.seq.replace('-','')}\n\n"
    output2 += f">{records.id}\n{records.seq.translate()}\n\n"
print(output1)
with open(f"{location}/{file_name}.aligned.fas.ungapped.fas", 'w') as out_file:
    out_file.write(output1)

with open(f"{location}/{file_name}.aligned.fas.translated.fas", 'w') as out_file:
    out_file.write(output2)

subprocess.run(f'cd {location}\ntrimal -colnumbering -htmlout trimal_output.html -in {file_name}.aligned.fas.translated.fas -out {file_name}.aligned.fas.trimal.fas -automated1 > columns_kept.txt', shell = True, stderr = subprocess.DEVNULL)
subprocess.run(f'cd {location}\ntrimal -backtrans {file_name}.aligned.fas.ungapped.fas -htmlout trimal_output.html -in {file_name}.aligned.fas.translated.fas -out {file_name}.aligned.fas.trimal.fas -automated1', shell = True, stderr = subprocess.DEVNULL)

subprocess.run(f'cd {location}\niqtree -s {file_name}.aligned.fas.trimal.fas -m MFP -t RANDOM -bb 1000 -alrt 1000 -bnni -nt AUTO', shell = True, stderr = subprocess.DEVNULL)
subprocess.run(f'cd {location}\niqtree -s {file_name}.aligned.fas.best.fas -m MFP -t RANDOM -bb 1000 -alrt 1000 -bnni -nt AUTO', shell = True, stderr = subprocess.DEVNULL)

subprocess.run(f'cd {location}\nhyphy busted --alignment {file_name}.aligned.fas.trimal.fas --tree {file_name}.aligned.fas.trimal.fas.treefile ', shell = True, stderr = subprocess.DEVNULL)
subprocess.run(f'cd {location}\nhyphy busted --alignment {file_name}.aligned.fas.best.fas --tree{file_name}.aligned.fas.best.fas.treefile ', shell = True, stderr = subprocess.DEVNULL)

```
