```python
location = "/mnt/c/Users/Saurav Baral/Desktop/20.RERConverge_trimmed/unTrimmed/1.Alignment"

import os
from Bio import SeqIO
list_of_files = os.listdir(location)
species_dictionary = {}
for i,file_name in enumerate(list_of_files):
    print (i, "of", len(list_of_files), end = ",")
    fasta_file = SeqIO.parse(f"{location}/{file_name}", "fasta")
    
    for records in fasta_file:
        species_dictionary.setdefault(records.id,0)
        species_dictionary[records.id] += 1

print(species_dictionary)
```

```python
output = "Species,Count"
for species, count in species_dictionary.items():
    output += f"\n{species},{count}" 
print(output)
with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/54.BUSCO_counts/counts.csv", "w") as out_file:
    out_file.write(output)
```
