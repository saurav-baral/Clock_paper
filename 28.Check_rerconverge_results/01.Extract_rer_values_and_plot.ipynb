```python
import svgwrite

# --- Input ---
names = species_new_list
highlighted = diapausing_species  # names to fill the rectangle

# --- Parameters ---
rect_width = 100
rect_height = 20
font_size = 14
padding = 10
row_height = rect_height + padding

# --- Create SVG ---
dwg = svgwrite.Drawing(f"{location}/28.Check_rerconverge_results/Negatives/{gene}.svg", profile="tiny",
                       size=(800, row_height * len(names) + padding))

for i, name in enumerate(names):
    y = i * row_height + padding
    # Write name
    dwg.add(dwg.text(name, insert=(10, y + rect_height - 5),
                     font_size=font_size, fill="black"))
    # Draw rectangle
    x_rect = 400  # position of rectangle to the right of name
    fill_color = "red" if name in highlighted else "blue"
    stroke_color = "black"
    dwg.add(dwg.rect(insert=(x_rect, y),
                     size=(rect_width, rect_height),
                     fill=fill_color,
                     stroke=stroke_color))

# --- Save file ---
dwg.save()
# print("{location}/28.Check_rerconverge_results/SVG saved as names.svg")
```
