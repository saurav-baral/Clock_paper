```python
import os
from Bio import SeqIO

location = "/mnt/griffin/saubar/Busco_runs_to_copy/"

list_of_genome_folders = os.listdir(location)
gene_count_dictionary = {}

for genome_name in list_of_genome_folders:
    list_of_gene_files = os.listdir(f'{location}/{genome_name}/CDS')
    for file_name in list_of_gene_files:
        gene_name = file_name.split(".")[0]
        if gene_name not in gene_count_dictionary:
            gene_count_dictionary[gene_name] = 1
        else:
            gene_count_dictionary[gene_name] += 1

output = 'Gene\tCount'

for key,value in gene_count_dictionary.items():
    output += f"\n{key}\t{value}"
    
with open("/mnt/griffin/saubar/Busco_gene_count.txt", 'w') as out_file:
    out_file.write(output)
    
```
