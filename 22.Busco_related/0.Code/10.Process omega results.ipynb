```python
import numpy as np


location = "/mnt/g/My Drive/Circadian Rhythm Genes Project"

output = "Gene\tOmega\tLower\tUpper"

with open(f"{location}/22.Busco_related/2.Omega_results/completed_omega", 'r') as omega_file:
    omega_list = omega_file.readlines()[1:]
gene_name_old = omega_list[0].split("\t")[0]
print(gene_name_old)
omega_value_list = []
for lines in omega_list:
    gene_name, branch, omega = lines.strip().split("\t")
    # print(gene_name)
    if gene_name != gene_name_old:
        
        median = np.median(omega_value_list)
        lower_quartile = np.percentile(omega_value_list, 25)
        upper_quartile = np.percentile(omega_value_list, 75)
        # print(gene_name_old,median, lower_quartile,upper_quartile)
        output += f"\n{gene_name_old}\t{median}\t{lower_quartile}\t{upper_quartile}"
        gene_name_old = gene_name
        omega_value_list = []

    omega_value_list.append(float(omega))    

with open(f"{location}/22.Busco_related/2.Omega_results/completed_omega_summarized", 'w') as out_file:
    out_file.write(output)
```
