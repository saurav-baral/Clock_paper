```python
from Bio import SeqIO

busco_gene_dic = {}

busco_fasta = SeqIO.parse("/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/4.Busco_aligned/1422at7088_nostop.fas.aligned.fas.best.fas", 'fasta')

for records in busco_fasta:
    busco_gene_dic[records.id] = records.seq.replace("-",'').upper()

tim_gene_dic = {}

tim_fasta = SeqIO.parse("/mnt/g/My Drive/Circadian Rhythm Genes Project/7.Timeless Exon Analysis/12.Family_specific_tests_timeless/Combined/Combined_timeless.fas", 'fasta')

for records in tim_fasta:
    tim_gene_dic[records.id.split("-")[0]] = records.seq.replace("-",'').upper()

# for key in tim_gene_dic:
#     print(key)
```

```python
import subprocess
for key in busco_gene_dic:
    if key in tim_gene_dic:
        seq1 = busco_gene_dic[key]
        seq2 = tim_gene_dic[key]
        # # print(seq1,seq2)
        # seq1 = "GTTTTTAGATATAGCA"
        # seq2 = "GTTAGATATAGCA"
        
        output = f">busco_{key}\n{seq1}\n>tim_{key}\n{seq2}"
        
        with open(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/5.Checking_busco_aligned/1.Timeless_1422at7088/{key}.txt",'w') as out_file:
            out_file.write(output)
        subprocess.run(f"cd '/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/5.Checking_busco_aligned/1.Timeless_1422at7088/'\n mafft --auto {key}.txt > {key}_aligned.txt", shell = True)
```

```python
import os
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/5.Checking_busco_aligned/1.Timeless_1422at7088/"
list_of_files  = os.listdir(location)
output = ''
for file_name in list_of_files:
    if file_name.endswith("_aligned.txt"):
        with open(f"{location}/{file_name}", 'r') as aligned_file:
            output += "".join(aligned_file.readlines())
with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/5.Checking_busco_aligned/1.Timeless_1422at7088.txt", 'w')             as out_file:
    out_file.write(output)
```
