```python
import os

location = "/mnt/griffin/saubar/Busco_runs_to_copy"

genome_location = "/mnt/griffin/saubar/Genomes_2023-12-26"


list_of_species_with_genome = os.listdir(genome_location)

genome_dictionary = {}

for species_name in list_of_species_with_genome:
    print(species_name)
    list_of_files_in_species_folder = os.listdir(f"{genome_location}/{species_name}")
    for file_name in list_of_files_in_species_folder:
        if file_name.endswith("_genomic.fna"):
            genome_dictionary[file_name] = species_name

list_of_genomes_with_busco = os.listdir(location)
output = 'Folder\tSpecies\tGene_count'
for folder_name in list_of_genomes_with_busco:
    genome_file_name = folder_name.replace("BUSCO_", '')
    print(genome_file_name)
    if genome_file_name in genome_dictionary:
        species  = genome_dictionary[genome_file_name]
        
        print(species)
        gene_count = len(os.listdir(f"{location}/{folder_name}/CDS"))
        output += f"\n{folder_name}\t{species}\t{gene_count}"
    else:
        print(genome_file_name, "Missing")

with open("/mnt/griffin/saubar/busco_result_summary.txt",'w') as out_file:
    out_file.write(output)
    


```
