```python
import os
from Bio import SeqIO

location = "/mnt/griffin/saubar/2.Individual_gene_files_BUSCO"

list_of_files = os.listdir(location)
counter = 0
output = ''
for file_name in list_of_files:
    fasta_file = SeqIO.parse(f"{location}/{file_name}", 'fasta')
    output_sequence = ''
    for records in fasta_file:
        extra = (len(records.seq)%3)
        if extra != 0:
            output_sequence += f">{records.id}\n{records.seq[:-extra]}\n"
        else:
            output_sequence += f">{records.id}\n{records.seq}\n"
        
    with open(f"/mnt/griffin/saubar/2.Individual_gene_files_BUSCO_by_3/{file_name}", 'w') as out_file:
        out_file.write(output_sequence)
        
#print(counter)

#with open(f"{location}/not_by_3.txt", 'w') as out_file:
#    out_file.write(output)
    
```
