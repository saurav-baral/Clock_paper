```python
import os

z_linked_list = [] 
with open("/mnt/griffin/saubar/1.Busco_results/z-linked_genes.txt", 'r') as in_file:
    for gene_name in in_file.readlines():
        z_linked_list.append(gene_name.strip())

list_of_files = os.listdir("/mnt/griffin/saubar/Individual_gene_files_BUSCO")
output = ''
new_z_list = []
for file_name in list_of_files:
    if file_name.split(".")[0] in z_linked_list:

        new_z_list.append(file_name.split(".")[0])
        output += f'{file_name.split(".")[0]}\n'
print(new_z_list, len(new_z_list))

with open("/mnt/griffin/saubar/1.Busco_results/z-linked_genes_filtered.txt", 'w') as out_file:
    out_file.write(output)
    
```
