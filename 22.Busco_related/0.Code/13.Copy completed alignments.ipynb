```python
from Bio import SeqIO
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related"

list_of_aligned_files = os.listdir(f"{location}/4.Busco_aligned")

if "desktop.ini" in list_of_aligned_files:
    list_of_aligned_files.remove("desktop.ini")
list_of_species = []
species_dictionary = {}
for files in list_of_aligned_files:
    print(files, end = ",")
    fasta_file = SeqIO.parse(f"{location}/4.Busco_aligned/{files}", 'fasta')
    for records in fasta_file:
        if records.id not in list_of_species:
            list_of_species.append(records.id)
            species_dictionary[records.id] = ''

print("\n\n")
for files in list_of_aligned_files:
    print(files, end = ",")
    current_gene_list = []
    fasta_file = SeqIO.parse(f"{location}/4.Busco_aligned/{files}", 'fasta') 
    for records in fasta_file:
        length_of_gene = len(records.seq)
        species_dictionary[records.id] += records.seq
        current_gene_list.append(records.id)
    if len(current_gene_list) != len(list_of_species):
        for species in list_of_species:
            if species not in current_gene_list:
                species_dictionary[species] += "N"*length_of_gene

output = ''
for key,value in species_dictionary.items():
    output += f">{key}\n{value}\n\n"

with open(f"{location}/merged.fas",'w')as out_file:
    out_file.write(output)


#\4.Busco_aligned
```
