```python
from Bio import Phylo
import os

list_of_files= os.listdir(f'{location}/4.domain_trees')
output = ''
for file_name in list_of_files:
    tree = Phylo.read(f'{location}/4.domain_trees/{file_name}', 'newick')
    species_list = [term.name for term in tree.get_terminals()]
    if (len(species_list))> 87:
        with open(f'{location}/4.domain_trees/{file_name}', 'r') as tree_file:
            
            output += f'{file_name}\t{tree_file.readlines()[0]}\n'
        
        # print(file_name,len(species_list))
with open(f'{location}/large_trees.tre', 'w') as domain_file:
    domain_file.write(output)
```
