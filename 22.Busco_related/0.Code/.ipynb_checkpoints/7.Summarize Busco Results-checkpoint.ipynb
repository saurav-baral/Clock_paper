```python
import os

with open("/mnt/griffin/saubar/busco_result_summary.txt", 'r') as result_summary_file:
    result_summary_list = result_summary_file.readlines()[1:]
name_dictionary = {}
for lines in result_summary_list:
    lines_split = lines.split("\t")
    name_dictionary[lines_split[0]] = lines_split[1] 


location = "/mnt/griffin/saubar/busco_runs_CDS_files"
list_of_genome_folders = os.listdir(location)

for folder_names in list_of_genome_folders:
    if folder_names.endswith("_genomic.fna"):
        with open(f"{location}/{folder_names}/run_lepidoptera_odb10/short_summary.txt", 'r') as summary_file:
            
            for lines in summary_file.readlines():
                if lines.startswith("	C:"):
                    print(name_dictionary[folder_names],lines)
                    break
            
```
