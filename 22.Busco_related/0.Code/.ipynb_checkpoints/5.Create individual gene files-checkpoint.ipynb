```python
import os
from Bio import SeqIO


with open("/mnt/griffin/saubar/busco_result_summary.txt", 'r') as result_summary_file:
    result_summary_list = result_summary_file.readlines()[1:]
name_dictionary = {}
for lines in result_summary_list:
    lines_split = lines.split("\t")
    name_dictionary[lines_split[0]] = lines_split[1] 




location = "/mnt/griffin/saubar/Busco_runs_to_copy/"

list_of_genome_folders = os.listdir(location)
gene_sequence_dictionary = {}

for genome_name in list_of_genome_folders:
    list_of_gene_files = os.listdir(f'{location}/{genome_name}/CDS')
    gene_sequence = ''
    for file_name in list_of_gene_files:
        fasta_file = SeqIO.parse(f'{location}/{genome_name}/CDS/{file_name}', 'fasta')
        for records in fasta_file:
            if "*" not in records.translate()[:-1]:
                gene_sequence = f">{name_dictionary[genome_name]}\n{records.seq}\n"
                gene_name = file_name.split(".")[0]
                if gene_name not in gene_sequence_dictionary:
                    gene_sequence_dictionary[gene_name] = gene_sequence
                else:
                    gene_sequence_dictionary[gene_name] += gene_sequence
        



for key,value in gene_sequence_dictionary.items():
    with open(f"/mnt/griffin/saubar/Individual_gene_files_BUSCO/{key}.fas", 'w') as out_file:
        out_file.write(value)
        
    
# with open("/mnt/griffin/saubar/Busco_gene_count.txt", 'w') as out_file:
#     out_file.write(output)
    
```
