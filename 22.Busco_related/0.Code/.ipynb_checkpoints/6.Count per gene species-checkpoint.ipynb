```python
import os
from Bio import SeqIO

list_of_gene_files = os.listdir("/mnt/griffin/saubar/Individual_gene_files_BUSCO")
output = "Gene\tCount"
for file_name in list_of_gene_files:
    fasta_file = SeqIO.parse(f"/mnt/griffin/saubar/Individual_gene_files_BUSCO/{file_name}", 'fasta')
    counter = 0
    for records in fasta_file:
        counter += 1
    output += f"\n{file_name.split('.')[0]}\t{counter}"

with open("/mnt/griffin/saubar/Busco_gene_count_filtered.txt", 'w') as out_file:
    out_file.write(output)

# with open()
```

```python
#for filtering

import subprocess

with open("/mnt/griffin/saubar/Busco_gene_count_filtered.txt", 'r') as count_file:
    count_list = count_file.readlines()[1:]

for line in count_list:
    gene, count = line.strip().split("\t")

    if int(count) < 70:
        subprocess.run(f'rm /mnt/griffin/saubar/Individual_gene_files_BUSCO/{gene}.fas', shell = True)
```
