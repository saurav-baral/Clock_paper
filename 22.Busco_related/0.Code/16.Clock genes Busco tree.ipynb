```python
from Bio import SeqIO
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related"

list_of_aligned_files = os.listdir(f"{location}/4.Busco_aligned")

if "desktop.ini" in list_of_aligned_files:
    list_of_aligned_files.remove("desktop.ini")
list_of_species = []

for files in list_of_aligned_files:
    # print(files, end = ",")
    fasta_file = SeqIO.parse(f"{location}/4.Busco_aligned/{files}", 'fasta')
    for records in fasta_file:
        if records.id not in list_of_species:
            list_of_species.append(records.id)
    if (len(list_of_species)) == 89:
        break
print(list_of_species) 
```

```python
from Bio import SeqIO
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/5.From_Busco/"

list_of_clock_genes = os.listdir(f"{location}/4.Clock_genes/1.Genes")

if "desktop.ini" in list_of_clock_genes:
    list_of_clock_genes.remove("desktop.ini")

for gene_file_name in list_of_clock_genes:
    output = ''
    fasta_file = SeqIO.parse(f"{location}/4.Clock_genes/1.Genes/{gene_file_name}", 'fasta')
    for records in fasta_file:
        if records.id in list_of_species:
            output += f">{records.id}\n{records.seq}\n"
        elif records.id == "Nymphalis_c":
            output += f">Nymphalis_c_album\n{records.seq}\n"
    with open(f"{location}/4.Clock_genes/2.Pruned_genes/{gene_file_name}", 'w') as out_file:
        out_file.write(output)
```

```python
from Bio import SeqIO
import os
from Bio import Phylo


location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/5.From_Busco/"

list_of_clock_genes = os.listdir(f"{location}/4.Clock_genes/2.Pruned_genes")

if "desktop.ini" in list_of_clock_genes:
    list_of_clock_genes.remove("desktop.ini")

for gene_file_name in list_of_clock_genes:
    fasta_file = SeqIO.parse(f"{location}/4.Clock_genes/2.Pruned_genes/{gene_file_name}", 'fasta')
    current_species_list  = []
    for records in fasta_file:
        current_species_list.append(records.id)
    print(len(current_species_list))
    tree = Phylo.read(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/5.From_Busco/combined_file.fas.treefile_no_branch_length.nwk", 'newick')
    species_list = [term.name for term in tree.get_terminals()]
    print(len(species_list))
    for species in species_list:
        if species not in current_species_list:
            print(species)
            tree.prune(species)
    Phylo.write(tree, f"/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/5.From_Busco/4.Clock_genes/3.Pruned_branches/{gene_file_name}.nwk", "newick")
    
```

```python
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/5.From_Busco/4.Clock_genes"

list_of_results = os.listdir(f"{location}/4.Hyphy_results")
if "desktop.ini" in list_of_results:
    list_of_results.remove("desktop.ini")
output = ''
for result_file in list_of_results:
    with open(f"{location}/4.Hyphy_results/{result_file}", 'r') as result_file_open:
        try:
            tree = result_file_open.readlines()[-1].split("=")[1]
        except:
            continue
    with open(f"{location}/5.Trees/{result_file.split('_')[1]}.txt", 'w') as out_file:
        out_file.write(tree)
    output += f"{result_file.split('_')[1]}\t{tree}"

with open(f"{location}/combined_tree.txt", 'w') as out_file:
    out_file.write(output)
```
