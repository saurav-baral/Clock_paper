```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/"
import statistics

domain_omega_dictionary = {}
domain_gene_dictionary = {}

with open(f"{location}/39.PfamDomain_search/2.Domain_info_pfam_scan/pfam_count.csv", "r") as domain_name_file:
    for domain_name_line in domain_name_file.readlines()[1:]:
        domain_name_split =domain_name_line.split(",") 
        domain_omega_dictionary[domain_name_split[0]] = domain_name_split[1]
gene_list = []
with open(f"{location}/39.PfamDomain_search/2.Domain_info_pfam_scan/domains.csv", 'r') as domain_file:
    for domain_line in domain_file.readlines()[1:]:
        lines_split = domain_line.split(",")
        gene_name = lines_split[0].split("_")[-1]
        if lines_split[5].split(".")[0] in domain_omega_dictionary:
            if lines_split[5].split(".")[0] not in domain_gene_dictionary:
                domain_gene_dictionary[lines_split[5].split(".")[0]] = []
            if gene_name not in gene_list:
                domain_gene_dictionary[lines_split[5].split(".")[0]].append(gene_name)
            gene_list.append(gene_name)

gene_omega_dic = {}

with open(f"{location}/35.Busco_all_genes/8.ABSREL_with-ClipKIT/all_omega.tsv", 'r') as omega_file:
    for omega_line in omega_file.readlines():
        omega_split = omega_line.strip().split(",")
        gene_name = omega_split[0]
        omega_list = []
        for omega in omega_split[1:]:
            omega_list.append(float(omega))
        gene_omega_dic[gene_name] = statistics.median(omega_list)


        
        
```

```python
output = "Domain_Name,Omega_value"
for key,value in domain_omega_dictionary.items():
    print("\n\n",key, len(set(domain_gene_dictionary[key])), len(domain_gene_dictionary[key]))
    for gene_name in set(domain_gene_dictionary[key]):
        try:
            output += f"\n{value} ({key}),{gene_omega_dic[gene_name]}"
        except:
            print(gene_name, "not found")
        # print(gene_omega_dic[gene_name])

with open(f"{location}/39.PfamDomain_search/2.Domain_info_pfam_scan/domain_omega.csv", 'w') as out_file:
    out_file.write(output)
```
