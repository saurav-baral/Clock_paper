```python
import os
from Bio import SeqIO

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project"

list_of_genes = os.listdir(f"{location}/22.Busco_related/2.Individual_gene_files_BUSCO_by_3")

list_of_genes.remove("desktop.ini")
output = ''
for gene_name in list_of_genes:
    print(gene_name, end = " ")
    fasta_file = SeqIO.parse(f"{location}/22.Busco_related/2.Individual_gene_files_BUSCO_by_3/{gene_name}", 'fasta')
    
    for records in fasta_file:
        if "Danaus" in records.id or "Bicyclus" in records.id:
            output += f">{records.id}_{gene_name.split('.')[0]}\n"
            translated_seq = records.seq.translate()
            if "*" in translated_seq[-1]:
                 output += translated_seq[:-1] +"\n"
            else:
                output += translated_seq + "\n"
            # print(output)
            break

with open(f"{location}/39.PfamDomain_search/1.Sequences/translated_sequence.fas", 'w') as out_file:
    out_file.write(str(output))
```
