```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/39.PfamDomain_search"

domain_count_dictionary = {}
gene_domain_dictionary = {}
pfam_name_dictionary = {}
old_gene_name = ''
with open(f"{location}/2.Domain_info_pfam_scan/domains.csv", 'r') as domain_file:
    for domain_line in domain_file.readlines()[1:]:
        lines_split = domain_line.split(",")
        gene_name = lines_split[0].split("_")[-1]
        if lines_split[6] != "WD40":
            if gene_name not in gene_domain_dictionary:
                gene_domain_dictionary[gene_name] = [lines_split[5].split(".")[0],lines_split[6],float(lines_split[12])]
            else:
                gene_domain_dictionary[f"{gene_name}_{len(gene_domain_dictionary[gene_name])}"] = [lines_split[5].split(".")[0],lines_split[6],float(lines_split[12])]
            
            if lines_split[5].split(".")[0] not in pfam_name_dictionary:
                pfam_name_dictionary[lines_split[5].split(".")[0]] = f"{lines_split[6]}_{lines_split[7]}"
            
for key,value in gene_domain_dictionary.items():
    gene_name  = key
    if "_" in key:
        gene_name = key.split("_")[0]
    if gene_name != old_gene_name:
        pfam_list = []
    if value[0] not in domain_count_dictionary:
        domain_count_dictionary[value[0]] = 0
    if value[0] not in pfam_list:
        domain_count_dictionary[value[0]] += 1
        pfam_list.append(value[0])
    old_gene_name = gene_name
count_list = []
output = "Pfam,Name,Count"
for key, value in domain_count_dictionary.items():
    # if value > 15:
        # output += f"\n{key},{pfam_name_dictionary[key]},{value}"
    count_list.append([value,key,pfam_name_dictionary[key]])
    # if value > 5:
    #     print(key,value)
# sorted(count_list, key=lambda x: x[0])
count_list = (sorted(count_list, key=lambda x: x[0], reverse=True))

for i in range(len(count_list)):
    if i < 20:
        output += f"\n{count_list[i][1]},{count_list[i][2]},{count_list[i][0]}"
print(output)
with open(f"{location}/2.Domain_info_pfam_scan/pfam_count.csv", 'w') as out_file:
    out_file.write(output)
```
