```python

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/35.Busco_all_genes/2.GSEA_RERConverge/1.GSEA_RERConverge_low_RHO"
gene_list = []

with open(f"{location}/candidate_gene.tsv", 'r') as gene_file:
    for gene_name in gene_file.readlines()[1:]:
        gene_list.append(gene_name.strip())
print(gene_list)    
```

```python
go_term = "GO:0051402"
go_gene_list = []
with open(f"{location}/GO_header.tsv", 'r') as go_head_file:
    for line in go_head_file.readlines()[1:]:
        gene_name = line.split("\t")[0]
        if  gene_name in gene_list:
            if go_term in line.split("\t")[1].strip().split(","):
               go_gene_list.append(gene_name)
print(go_gene_list)   
```
