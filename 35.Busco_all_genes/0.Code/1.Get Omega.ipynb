```python
import os
import json

location = "/mnt/d/Circadian Rhythm Genes Project_Large_Files/35.Busco_all_genes/8.ABSREL_with-ClipKIT/complete_set1"

list_of_folders= os.listdir(location)
# print(list_of_folders)
# list_of_folders = ["6at7088"]
output = ''
for i,folder_name in enumerate(list_of_folders):
    if "." not in folder_name:
        print(f"{i},{int(len(list_of_folders))}", end = "  ")
        list_of_files = os.listdir(f"{location}/{folder_name}")
        for file_name in list_of_files:
            if file_name.endswith("ABSREL.json"):
                
                output += f"\n{folder_name.split('_')[0]}"
                with open(f"{location}/{folder_name}/{file_name}", 'r') as json_data:
                    json_file= json.load(json_data)
                    for key,value in json_file["branch attributes"]["0"].items():
                        omega = 0
                        for value in ((json_file["branch attributes"]["0"][key]["Rate Distributions"])):
                            omega += float(value[0])*float(value[1])
                        output += f",{omega}"
                        # print(",",omega, end = "   ")
                        
                    # for key in json_file["branch attributes"]["0"]:
                        # print(key)
                        # assert False
        # print(list_of_files)
        # break
with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/35.Busco_all_genes/8.ABSREL_with-ClipKIT/all_omega.tsv", 'w') as out_file:
    out_file.write(output)
```
