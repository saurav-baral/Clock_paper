```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/35.Busco_all_genes/11.GSEA_ERC_top_100"

import os


list_of_files = os.listdir(location)
for file_name in list_of_files:
    if file_name.startswith("GSEA_"):
        with open(f"{location}/{file_name}/candidate_gene_100.csv.BP.GSEA_result.tsv", 'r') as gsea_file:
            output = "Enriched Ontological Terms,Enrichment Score,p-value,Number of Genes"
            for lines in gsea_file.readlines()[1:11]:
                lines_split = lines.split("\t")
                # print(lines_split)
                go_term = f"{lines_split[1]} ({lines_split[0]})"
                en_score = round(float(lines_split[3])/float(lines_split[4]),2)
                p_value = lines_split[9].strip()
                N_genes = lines_split[3]
                output += f"\n{go_term},{en_score},{p_value},{N_genes}"
                # print(output)
                # break
                # 
                # print(lines)
        print(output)
        with open(f"{location}/{file_name}/plot_gsea.csv",'w') as out_file:
            out_file.write(output)
                
            # print(gsea_file.readlines())
```
