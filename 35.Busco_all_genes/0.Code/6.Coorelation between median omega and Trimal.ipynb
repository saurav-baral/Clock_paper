```python
import numpy as np
import matplotlib.pyplot as plt
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/35.Busco_all_genes"

# Step 1: Load data from a CSV file
# Assuming the CSV has two columns: 'x' and 'y' (without headers)
data = np.loadtxt(f'{location}/Median_omega_percentage.csv', delimiter=',', skiprows=1,usecols=(1, 2))  # Update `skiprows` if there are headers
x = data[:, 0]  # First column
y = data[:, 1]  # Second column

# Step 2: Calculate Pearson correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Step 3: Calculate R-squared
r_squared = correlation ** 2

# Step 4: Plot the data
plt.scatter(x, y, color='blue', label=f'Correlation: {correlation:.2f}')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red', label=f'R: {r_squared:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Correlation')
plt.legend()
plt.grid(True)
plt.show()

# Print results
print("Correlation coefficient (r):", correlation)
print("Coefficient of determination (R):", r_squared)
```
