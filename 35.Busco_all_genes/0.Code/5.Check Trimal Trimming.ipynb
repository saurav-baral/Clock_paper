```python
from Bio import SeqIO
import os

location = "/mnt/c/Users/Saurav Baral/Desktop/Busco_related/3.Busco_processing_individual_genes"

list_of_folders= os.listdir(location)
# list_of_folders = ["6at7088"]
output = 'Gene\tPercenage_kept'
for i,folder_name in enumerate(list_of_folders):
    if "." not in folder_name:
        # print(folder_name)
        print(f"{i},{int(len(list_of_folders))}", end = "   ")
        list_of_files = os.listdir(f"{location}/{folder_name}")
        for file_name in list_of_files:
            
            if file_name.endswith("_nostop.fas.aligned.fas.best.fas"):
                
                without_trimming = SeqIO.parse(f"{location}/{folder_name}/{file_name}", 'fasta')
                for records in without_trimming:
                    original_length = len(records.seq)
                    break
            if file_name.endswith("_nostop.fas.aligned.fas.trimal.fas"):
                with_trimming = SeqIO.parse(f"{location}/{folder_name}/{file_name}", 'fasta')
                for records in with_trimming:
                    new_length = len(records.seq)
                    break
        percentage_kept = new_length/original_length
        # print(new_length,original_length)
        # print(percentage_kept)
        # assert False
        output += f"\n{folder_name}\t{percentage_kept}"
with open(f"{location}/Trimal_percentage.txt", 'w') as out_file:
    out_file.write(output)
```
