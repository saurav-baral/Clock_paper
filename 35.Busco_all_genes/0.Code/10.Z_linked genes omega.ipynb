```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/"
import statistics

gene_omega_dic = {}

with open(f"{location}/35.Busco_all_genes/8.ABSREL_with-ClipKIT/all_omega.tsv", 'r') as omega_file:
    for omega_line in omega_file.readlines():
        omega_split = omega_line.strip().split(",")
        gene_name = omega_split[0]
        omega_list = []
        for omega in omega_split[1:]:
            omega_list.append(float(omega))
        gene_omega_dic[gene_name] = statistics.median(omega_list)
```

```python
output = "Gene,Omega"
with open(f"{location}/35.Busco_all_genes/9.Z-linked/z-linked_genes.txt",'r') as zlinked_files:
    for gene_name in zlinked_files.readlines():
        try:
            print(gene_omega_dic[gene_name.strip()])
        
            output += f"\n{gene_name.strip()},{gene_omega_dic[gene_name.strip()]}"
        except:
            print(gene_name.strip())
with open(f"{location}/35.Busco_all_genes/9.Z-linked/omega_values.csv",'w') as out_file:
    out_file.write(output)
```
