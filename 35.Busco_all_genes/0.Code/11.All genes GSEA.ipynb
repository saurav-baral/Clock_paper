```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/35.Busco_all_genes/11.GSEA_ERC_top_100"

import os

list_of_genes = []
list_of_files = os.listdir(location)
for file_name in list_of_files:
    if file_name.startswith("GSEA_"):
        with open(f"{location}/{file_name}/candidate_gene_100.csv", 'r') as gsea_file:
            for gene_name in gsea_file.readlines()[1:]:
                if gene_name.strip() not in list_of_genes:
                    list_of_genes.append(gene_name.strip())

with open(f"{location}/all_candidate_gene_100.csv", 'w') as gsea_file:
    gene_list = '\n'.join(list_of_genes)
    output = f"geneid\n{gene_list}"
    gsea_file.write(output)    
```
