```python
location = "/mnt/c/Users/Saurav Baral/Desktop/resuts"
import os
import json
import statistics
list_of_files = os.listdir(f"{location}")
# list_of_files.remove("desktop.ini")

omega_dictionary = {}
for file_name in list_of_files:
    with open(f"{location}/{file_name}", 'r') as json_data:
        json_file= json.load(json_data)
        for key,value in json_file["branch attributes"]["0"].items():
            omega = 0
            for value in ((json_file["branch attributes"]["0"][key]["Rate Distributions"])):
                omega += float(value[0])*float(value[1])
            gene_name = file_name.split("_")[0]
            if gene_name not in omega_dictionary:
                omega_dictionary[gene_name] = []
            omega_dictionary[gene_name].append(omega)

print(omega_dictionary)
output = "Gene,Median_omega"
for key, value in omega_dictionary.items():
    output += f"\n{key},{statistics.median(value)}"
with open(f"{location}/omega_absrel.txt", 'w') as out_file:
    out_file.write(output)
```

```python
output = "Gene,With_Trim,Without_Trim"
import statistics
with open(f"{location}/Top_5_percentage.csv",'r') as csv_file:
    for lines in csv_file.readlines()[1:]:
        lines_split = lines.strip().split(",")
        if lines_split[0] in omega_dictionary:
            output+= f"\n{lines_split[0]},{lines_split[1]},{statistics.median(omega_dictionary[lines_split[0]])}"
with open(f"{location}/Top_5_percentage_compared.csv",'w') as out_file:
    out_file.write(output)
```
