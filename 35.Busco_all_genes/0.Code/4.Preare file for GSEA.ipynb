```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project"

output = "Parent\tGO_term"

with open(f"{location}/22.Busco_related/1.Annotation/MM_xnpmgii0.emapper.annotations.tsv", 'r') as annotation_file:
    for gene_line in annotation_file.readlines():
        if "#" not in gene_line:
            gene_line_split  = gene_line.split("\t")
            output += f"\n{gene_line_split[0]}\t{gene_line_split[9]}"

with open(f"{location}/35.Busco_all_genes/GO_header.tsv", 'w') as out_file:
    out_file.write(output)
```
