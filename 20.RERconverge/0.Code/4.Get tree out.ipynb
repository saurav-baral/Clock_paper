```python
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/4.HYPHY branch length optimization"

list_of_folders = os.listdir(f"{location}/3.Std_out")

if "desktop.ini" in list_of_folders:
    list_of_folders.remove("desktop.ini")
output = ''
for file in list_of_folders:
    with open(f"{location}/3.Std_out/{file}", 'r') as std_file:
        file_content = std_file.readlines()
    tree = ''
    for lines in file_content:
        if "Tree givenTree" in lines:
            tree = lines.split("=")[1]
            output += f"{file.split('.')[0]}\t{tree}\n"
    
    with open(f"{location}/4.Optimized_tree/{file}", 'w') as out_file:
        out_file.write(tree)
with open(f"{location}/combined_tree.nwk", 'w') as out_file:
    out_file.write(output)
```
