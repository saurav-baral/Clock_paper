```python
from Bio import SeqIO
from Bio import Phylo
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/20.RERconverge/4.HYPHY branch length optimization"
tree = Phylo.read(f"{location}/combined_file.fas.treefile_no_branch_length.nwk", "newick")
species_list = [term.name for term in tree.get_terminals()]
list_of_folders = os.listdir(f"{location}/1.Gene_sequences")
if "desktop.ini" in list_of_folders:
    list_of_folders.remove("desktop.ini")

for files in list_of_folders:
    species_gene_specific = []
    tree = Phylo.read(f"{location}/combined_file.fas.treefile_no_branch_length.nwk", "newick")
    fasta_file = SeqIO.parse(f"{location}/1.Gene_sequences/{files}", "fasta")
    for records in fasta_file:
        
        species_gene_specific.append(records.id)
    for species in species_list:
        if species not in species_gene_specific:
            print(files, species)
            tree.prune(species)
    Phylo.write(tree, f"{location}/2.Pruned_branches/{files}.nwk", "newick")

# print(species_list)

```
