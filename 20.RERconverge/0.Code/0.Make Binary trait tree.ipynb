```python
from Bio import Phylo
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project"
with open(f"{location}/20.RERconverge/1.Diapause_data/Two_state.txt", 'r') as state_file:
    state_list = state_file.readlines()
diapause_state_dic = {}
for line in state_list:
    species, state = line.strip().split("\t")
    diapause_state_dic[species.split("-")[0]] = state

# print(diapause_state_dic)
tree = Phylo.read(f"{location}/20.RERconverge/3.HYPHY branch length optimization/4.HYPHY_trees/cad.nwk", 'newick')
species_list = [term.name for term in tree.get_terminals()]
for species in species_list:
    if species not in diapause_state_dic:
        print(species)

for clade in tree.find_clades():
    if clade.name in diapause_state_dic:
        if diapause_state_dic[clade.name] == "diapause":
            # print(clade)
            clade.branch_length = 1
        else:
            clade.branch_length = 0
    else:
        clade.branch_length = 0
Phylo.write(tree, f"{location}/20.RERconverge/1.Diapause_data/Two_state.nwk", "newick")
    # print(clade)
# print(species_list)
```
