```python
from Bio import SeqIO
import os
import subprocess

family_group = "4.Pierinae"
source_location = "/mnt/k/My Drive/MRJP/3.Phylo/3.aligned"
query_species = "Pieris_brassicae"


working_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}"
output_location = f"{working_location}/1.Query"
list_of_working_folders = os.listdir(working_location)
if "1.Query" in list_of_working_folders:
    subprocess.run(f'rm -r "{output_location}"', shell = True)
os.mkdir(output_location)



list_of_genes = os.listdir(source_location)
if "desktop.ini" in list_of_genes:
    list_of_genes.remove("desktop.ini")
output = ""
for gene_file in list_of_genes:
    

    source_file = SeqIO.parse(f"{source_location}/{gene_file}", "fasta")

    for species_name in source_file:
        if query_species in species_name.id:
            gene_name = species_name.id.split("_")[-2]
            print(species_name.id, gene_name)
            
            translated_sequence = (species_name.seq.translate()).replace("-","")
    
            output += f">{query_species}_{gene_name}\n{translated_sequence}\n\n"
print(output)
with open(f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}/1.Query/query.txt", 'w') as query_out:
    query_out.write(output)
```
