```python
import os
from Bio import SeqIO
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny"
family_group_list = ["3.Satyrine","4.Pierinae","5.Coliadinae","6.Heliconiinae_Danainae_Nymphalinae","7.Papilionidae","8.Lycaenidae"]

gene_list = os.listdir(f"{location}/3.Satyrine/3.Extracted Sequences/")
if "desktop.ini" in gene_list:
    gene_list.remove("desktop.ini")

for gene_name in gene_list:
    print(gene_name)
    output = ''
    for family_group in family_group_list:
        print(family_group)
        list_of_species_files = os.listdir(f"{location}/{family_group}/3.Extracted Sequences/{gene_name}")
        if "desktop.ini" in list_of_species_files:
            list_of_species_files.remove("desktop.ini")
        for species_name in list_of_species_files:
            # print(species_name)
            fasta_file= SeqIO.parse(f"{location}/{family_group}/3.Extracted Sequences/{gene_name}/{species_name}", 'fasta')
            for sequences in fasta_file:
                output += f">{sequences.id.replace(gene_name, '')[:-1]}\n{sequences.seq}\n"
    with open(f"{location}/10.All_gene_sequences/{gene_name}.fas", 'w') as out_file:
        out_file.write(output)
    # print(output)
    # break
```

```python
import os
from Bio import SeqIO
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny"

list_of_files = os.listdir(f"{location}/10.All_gene_sequences")
list_of_species = []
aligned_files = []
for file_name in list_of_files:
    if file_name.endswith("aligned.fas.best.fas"):
        # print(file_name)
        aligned_files.append(file_name)
        fasta_file = SeqIO.parse(f"{location}/10.All_gene_sequences/{file_name}", "fasta")
        for records in fasta_file:
            if records.id not in list_of_species:
                list_of_species.append(records.id)
output = ''
for species in list_of_species:
    full_sequence = ''
    for file_name in aligned_files:
        sequence = ''
        sequence_length = ''
        fasta_file = SeqIO.parse(f"{location}/10.All_gene_sequences/{file_name}", "fasta")
        for records in fasta_file:
            sequence_length = len(records.seq)
            if records.id == species:
                sequence = records.seq
        if sequence == '':
            print(file_name,"\t\t",species, "Missing")
            sequence = "N"*sequence_length
        full_sequence += sequence
    print(species, len(full_sequence))
    output += f">{species}\n{full_sequence}\n\n"

with open(f"{location}/11.Combined_again/combined_file.fas", 'w') as out_file:
    out_file.write(output)
        
    
```
