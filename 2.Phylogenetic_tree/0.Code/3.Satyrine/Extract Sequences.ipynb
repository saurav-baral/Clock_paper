```python
def get_genome_file(species):
    list_of_genomes = os.listdir("/mnt/f/Genomes_2023-12-26")
    for genome in list_of_genomes:
        if species == genome:
            files_in_genome_folder = os.listdir(f"/mnt/f/Genomes_2023-12-26/{genome}")
            for file in files_in_genome_folder:
                if file.endswith("_genomic.fna"):
#                     print(file)
                    return(file)
#                     break
    
```

```python
import os
import subprocess
from Bio import SeqIO


family_group = "3.Satyrine"


working_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}"
blast_output_location = f"{working_location}/1.Blast_result"
output_location = f"{working_location}/3.Extracted Sequences"

list_of_working_folders = os.listdir(working_location)
if "3.Extracted Sequences" in list_of_working_folders:
    subprocess.run(f'rm -r "{output_location}"', shell = True)
os.mkdir(output_location)


list_of_species = os.listdir(blast_output_location)

if "desktop.ini" in list_of_species:
    list_of_species.remove("desktop.ini")
    
print(list_of_species)
list_of_errors = ''
for species in list_of_species:
    print(species)
    with open(f"{blast_output_location}/{species}/{species}_coordinates.csv", 'r') as coordinates_file:
        coordinates_list = coordinates_file.readlines()
        
    for i in range(1,len(coordinates_list)):
        coordinates_split = coordinates_list[i].split(",")
        
        gene = coordinates_split[6].split("_")[-1]
        
        
        
        scaffold,start,stop,complement = coordinates_split[1],coordinates_split[2],coordinates_split[3],coordinates_split[4]
#         genome_file = get_genome_file(species)
        
        genome_fasta = SeqIO.parse(f"/mnt/f/Genomes_2023-12-26/{species}/{get_genome_file(species)}", 'fasta')
        
        for records in genome_fasta:
            if scaffold in records.id:
                sequence = (records.seq[int(start)-1:int(stop)])
                break
        if complement == "1":
            sequence = sequence.reverse_complement()
            
        
        print(gene)
#         print(sequence,"\n")
        if "*" in sequence.translate():
            list_of_errors += f"{species}\t{gene}\n"
            print(sequence.translate())
            continue
            
        list_of_folders_in_extracted = os.listdir(f"{output_location}")
        
        if gene not in list_of_folders_in_extracted:
            os.mkdir(f"{output_location}/{gene}")
        
        print(list_of_errors)  
        with open(f"{output_location}/{gene}/{species}.txt", 'w') as out_file:
            output = f">{species}_{gene}\n{sequence}"
            out_file.write(output)
        
#         print(coordinates_list[i])
#     assert False

with open(f"{working_location}/errors.txt", 'w') as error_file:
    error_file.write(list_of_errors)
```
