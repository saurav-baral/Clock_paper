```python
import os
import re
import io
import subprocess
```

```python
def get_genome_file(genome_location,species):
    list_of_files_in_genome_folder = os.listdir(f"{genome_location}/{species}")
    for file in list_of_files_in_genome_folder:
        if file.endswith("_genomic.fna"):
            genome_file = file
    return(genome_file)
```

```python
# list_of_species = os.listdir("/mnt/h/My Drive/Circadian Rhythm Genes Project/6.Period Exon Analysis/1.Annotated Species")
list_of_species = ["Baronia_brevicornis","Battus_philenor","Iphiclides_podalirius","Luehdorfia_chinensis","Ornithoptera_alexandrae","Ornithoptera_priamus","Papilio_dardanus_tibullus","Papilio_demoleus","Papilio_elwesi","Papilio_gigon","Papilio_glaucus","Papilio_machaon","Papilio_memnon","Papilio_polytes","Papilio_protenor","Papilio_xuthus","Parnassius_glacialis","Sericinus_montela","Teinopalpus_imperialis","Troides_aeacus","Troides_oblongomaculatus"]
query_species = "Papilio_xuthus"
family_group = "7.Papilionidae"
query_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}/1.Query"

query_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}/1.Query/query.txt"
list_of_genomes = os.listdir("/mnt/f/Genomes_2023-12-26")
```

```python
working_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}"
output_location = f"{working_location}/1.Blast_result"
list_of_working_folders = os.listdir(working_location)
if "1.Blast_result" in list_of_working_folders:
    subprocess.run(f'rm -r "{output_location}"', shell = True)
os.mkdir(output_location)




genome_file = ''
output = ''
for species in list_of_species:
    # if "
    # species = species_folder.split(".")[1]
    blast_output_location = f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}/1.Blast_result"
    blast_output_folders = os.listdir(blast_output_location)
    if species not in blast_output_folders:
        os.mkdir(f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}/1.Blast_result/{species}")
    
    
    for genome in list_of_genomes:
        if species == genome:
            files_in_genome_folder = os.listdir(f"/mnt/f/Genomes_2023-12-26/{genome}")
            for file in files_in_genome_folder:
                if re.search('_genomic.fna$', file):
                    genome_file = file
                    break
    
    output += f'echo {species}\ntblastn -seg no -query "{query_location}" -db "/mnt/f/Genomes_2023-12-26/{species}/{genome_file}" -num_alignments 3 -out "{blast_output_location}/{species}/{species}_blast_out.htm" -html \n\n tblastn -seg no -query "{query_location}" -db "/mnt/f/Genomes_2023-12-26/{species}/{genome_file}" -num_alignments 3 -out "{blast_output_location}/{species}/{species}_blast_out.txt"\n\n'
    
with io.open(f"/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/{family_group}/tblastn_{query_species}.sh", "w", newline='\n') as outfile:
    outfile.write(output)
#             print("here",genome)
#     print(species_folder)

```
