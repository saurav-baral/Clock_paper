```python
location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/"

import os

list_of_family_groups = ["3.Satyrine","4.Pierinae","5.Coliadinae","6.Heliconiinae_Danainae_Nymphalinae","7.Papilionidae","8.Lycaenidae"]

list_of_genes = os.listdir(f'{location}/{list_of_family_groups[0]}/3.Extracted Sequences')

if "desktop.ini" in list_of_genes:
    list_of_genes.remove("desktop.ini")
                           
for gene in list_of_genes:
    output = ''
    for family_group in list_of_family_groups:
        list_of_species = os.listdir(f"{location}/{family_group}/1.Blast_result")

        with open(f"{location}/{family_group}/errors.txt",'r') as error_file:
            error_list = error_file.readlines()

        for error_species in error_list:

            error_species_name = error_species.split("\t")[0]
            if error_species_name in list_of_species:
                list_of_species.remove(error_species_name)


        if "desktop.ini" in list_of_species:
            list_of_species.remove("desktop.ini")

        for species in list_of_species:
            with open(f"{location}/{family_group}/3.Extracted Sequences/{gene}/{species}.txt", 'r') as gene_sequence:
                output += "\n"+ "".join(gene_sequence.readlines())
    with open(f"{location}/1.Combined/{gene}.fas",'w') as out_file:
        out_file.write(output)
#     print(output)
#     assert False
#     print(family_group, list_of_species)
```
