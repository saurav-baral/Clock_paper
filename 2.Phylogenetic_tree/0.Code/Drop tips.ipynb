```python
from Bio import Phylo

# Load your tree (replace with your file)
tree = Phylo.read("your_tree_file.nwk", "newick")

# List of species you want to retain
keep_species = {"species1", "species2", "species3"}

# Iterate over terminals (tips) and prune those not in the list
for tip in tree.get_terminals():
    if tip.name not in keep_species:
        tree.prune(tip)

# Optional: Save the pruned tree
Phylo.write(tree, "pruned_tree.nwk", "newick")
```
