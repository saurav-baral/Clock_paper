```python
location = "/mnt/h/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/"

import os
from Bio import SeqIO


list_of_genes = os.listdir(f'{location}/3.Satyrine/3.Extracted Sequences')

if "desktop.ini" in list_of_genes:
    list_of_genes.remove("desktop.ini")

species_dictionary = {}
    
sequence_file = SeqIO.parse(f"{location}/1.Combined/{list_of_genes[0]}_aligned.fas",'fasta')

for gene in list_of_genes:
    print(gene)
    sequence_file = SeqIO.parse(f"{location}/1.Combined/{gene}_aligned.fas",'fasta')
    for records in sequence_file:
        species_name = ("_".join(records.id.split("_")[:-1]))
        if species_name not in species_dictionary:
            species_dictionary[species_name] = ""
        species_dictionary[species_name] += records.seq
output = ''
for key, value in species_dictionary.items():
    output += f">{key}\n{value}\n"

with open(f'{location}/2.Tree/combined_aligned.fas','w') as out_file:
    out_file.write(output)
```
