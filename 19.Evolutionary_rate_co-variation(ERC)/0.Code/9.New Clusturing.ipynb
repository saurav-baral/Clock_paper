```python
from collections import defaultdict
from itertools import combinations


def find_clusters(data):
    # Step 1: Create a graph based on common members
    graph = defaultdict(set)

    # Compare all pairs of keys
    for key1, key2 in combinations(data.keys(), 2):
        # Check if the two lists have common members
        if set(data[key1]) & set(data[key2]):  # Non-empty intersection
            graph[key1].add(key2)
            graph[key2].add(key1)

    # Step 2: Find connected components in the graph
    visited = set()
    clusters = []

    def dfs(node, cluster):
        visited.add(node)
        cluster.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor, cluster)

    for key in data.keys():
        if key not in visited:
            cluster = set()
            dfs(key, cluster)
            clusters.append(cluster)
            print(clusters)

    return clusters

# Call the function
clusters = find_clusters(new_dic)
print(clusters)
```

```python
import networkx as nx
from sknetwork.clustering import Louvain
import scipy.sparse as sp
import matplotlib.pyplot as plt

# Sample dictionary: key is gene, value is list of genes close to the key gene
gene_dict = {
    "GeneA": ["GeneB", "GeneC", "GeneD"],
    "GeneB": ["GeneA", "GeneC", "GeneE"],
    "GeneC": ["GeneA", "GeneB", "GeneF"],
    "GeneD": ["GeneA", "GeneE"],
    "GeneE": ["GeneB", "GeneD", "GeneF"],
    "GeneF": ["GeneC", "GeneE"]
}

# 1. Construct graph
G = nx.Graph()

# Add edges based on co-clustering (shared genes in each other's list)
for gene, close_genes in gene_dict.items():
    for close_gene in close_genes:
        if not G.has_edge(gene, close_gene):
            G.add_edge(gene, close_gene)

# 2. Convert the NetworkX graph to a scipy sparse matrix (CSR format)
adj_matrix = nx.to_scipy_sparse_matrix(G, nodelist=list(G.nodes()))
adj_matrix = sp.csr_matrix(adj_matrix)  # Ensure CSR format

# 3. Apply Louvain community detection
louvain = Louvain()
labels = louvain.fit_predict(adj_matrix)

# 4. Display the clustering result
# Each gene will be assigned a cluster (community) number
print("Gene clusters based on Louvain community detection:")
for i, gene in enumerate(G.nodes()):
    print(f"{gene}: Cluster {labels[i]}")

# Optionally, visualize the graph with clusters (requires matplotlib)
colors = labels  # Color nodes based on their community

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color=colors, cmap=plt.cm.rainbow, node_size=500, font_size=12)
plt.title("Gene Clusters (Louvain Algorithm)")
plt.show()
```

```python
import numpy as np

# 3. Simple modularity-based community detection (basic greedy approach)
def detect_communities(adj_matrix):
    # Initialize each node in its own community
    communities = {i: [i] for i in range(adj_matrix.shape[0])}
    
    # We will use a basic modularity approach: merge communities with the highest modularity gain
    def modularity_gain(community1, community2):
        # Simplified modularity calculation for merging two communities
        subgraph = adj_matrix[community1, :][:, community2]  # Submatrix for the two communities
        total_edges = subgraph.sum()
        expected_edges = len(community1) * len(community2) / adj_matrix.shape[0]
        return total_edges - expected_edges
    
    # Merge communities until we reach a point where no further gain is possible
    while len(communities) > 1:
        best_gain = 0
        best_communities = None
        
        for c1, community1 in communities.items():
            for c2, community2 in communities.items():
                if c1 != c2:
                    gain = modularity_gain(community1, community2)
                    if gain > best_gain:
                        best_gain = gain
                        best_communities = (c1, c2)
        
        # Merge the best communities
        if best_communities:
            c1, c2 = best_communities
            communities[c1].extend(communities.pop(c2))
        
    return list(communities.values())

# Apply the community detection
detected_communities = detect_communities(adj_matrix)

# Display the detected communities
print("Detected Communities:")
for i, community in enumerate(detected_communities):
    print(f"Community {i+1}: {community}")
```
