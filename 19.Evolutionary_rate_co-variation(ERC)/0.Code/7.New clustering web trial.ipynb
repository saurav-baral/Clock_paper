```python
import pandas as pd
import networkx as nx

# Example DataFrame
data = {
    'A': [0.0, 0.5, 0.2, 0.8],
    'B': [0.5, 0.0, 0.6, 0.3],
    'C': [0.2, 0.6, 0.0, 0.7],
    'D': [0.8, 0.3, 0.7, 0.0]
}
df = csv_file

# Threshold for clustering
threshold = 0.4
# n = 1000
# random_rows = df.sample(n, axis=0)

# # Sampling 100 random columns
# df = random_rows.sample(n, axis=1)


# Create a graph
G = nx.Graph()

# Add edges where the value is above the threshold
for i in df.index:
    print(i, end = ",")
    for j in df.columns:
        if i != j and df.loc[i, j] > threshold:  # Avoid self-loops
            G.add_edge(i, j)

# Find connected components (clusters)
clusters = list(nx.connected_components(G))
num_clusters = len(clusters)

print(f"Number of clusters: {num_clusters}")
print(f"Clusters: {clusters}")
```

```python
pos = nx.spring_layout(G)  # Position the nodes using a spring layout
plt.figure(figsize=(8, 6))
nx.draw(
    G, pos, with_labels=True, node_color="skyblue", 
    node_size=1500, edge_color="gray", font_weight="bold"
)

# Add edge labels (weights)
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels={(i, j): f"{w:.2f}" for (i, j), w in edge_labels.items()})

plt.title("Web-Like Clustering Based on Matrix Values")
plt.show()
```
