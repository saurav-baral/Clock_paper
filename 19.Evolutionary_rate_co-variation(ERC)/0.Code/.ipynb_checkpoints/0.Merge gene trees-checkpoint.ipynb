```python
from Bio import Phylo
gene_folder_list = ["6.Period Exon Analysis","7.Timeless Exon Analysis","9.CLOCK Exon Analysis","10.Cycle_Exon_Analysis","11.CWO_Exon_Analysis","12.Cryptochrome_1a_Exon_Analysis","13.Cryptochrome_1b_Exon_Analysis","14.Cryptochrome_2_Exon_Analysis","15.Timeout_Exon_Analysis"]
gene_list = ["period","timeless","CLOCK","Cycle","CWO","cry1a","cry1b","cry2","timeout"]
output = ''
for gene_folder, gene in zip(gene_folder_list,gene_list):
    # print(gene_folder, gene)
    location = f"/mnt/g/My Drive/Circadian Rhythm Genes Project/{gene_folder}/12.Family_specific_tests_{gene}/Nymphalidae"

    tree = Phylo.read(f"{location}/Nymphalidae_{gene.lower()}_nostop.fas.aligned.fas.trimal.fas.treefile", 'newick')
    species_list = [term.name for term in tree.get_terminals()]
    with open(f"{location}/Nymphalidae_{gene}_nostop.fas.aligned.fas.trimal.fas.treefile", 'r') as tree_file_open:
        tree_file = tree_file_open.readlines()[0]
        # print(tree_file)
    for species_name in species_list:
        tree_file = tree_file.replace(species_name, species_name.split("-")[0])
    output += f"{gene}\n{tree_file}\n"
print(output)
with open(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/combined_tree.nwk",'w') as out_file:
    out_file.write(output)
```
