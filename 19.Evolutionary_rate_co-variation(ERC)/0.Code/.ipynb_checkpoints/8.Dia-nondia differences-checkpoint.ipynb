```python
gene_dictionary_dia = {}

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/2.Two_state"

with open(f"{location}/ERC_correlation_diapause.csv", 'r') as cluster_file:
    cluster_lines = cluster_file.readlines()

genes = cluster_lines[0].replace("_nostop.fas.aligned.fas.best.fas.ClipKIT", '').replace('.fas.ClipKIT',"").strip().split(',')
# print(cluster_lines[1])

for gene_name in genes:
    gene_dictionary_dia[gene_name] = {}
    # print(gene_name)
for j,lines in enumerate(cluster_lines[1:]):
    print(j, end = ",")
    coorelations = lines.strip().split(',')
    for i,cor_val in enumerate(coorelations):
        gene_dictionary_dia[genes[j]][genes[i]] = [cor_val]   
```
