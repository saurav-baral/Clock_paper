```python
from Bio import SeqIO
from Bio import Phylo

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/"
gene_folder_list = ["6.Period Exon Analysis","7.Timeless Exon Analysis","9.CLOCK Exon Analysis","10.Cycle_Exon_Analysis","11.CWO_Exon_Analysis","12.Cryptochrome_1a_Exon_Analysis","13.Cryptochrome_1b_Exon_Analysis","14.Cryptochrome_2_Exon_Analysis","15.Timeout_Exon_Analysis"]
gene_list = ["period","timeless","CLOCK","Cycle","CWO","cry1a","cry1b","cry2","timeout"]

print(tree_species_list)
for gene_folder, gene in zip(gene_folder_list,gene_list):
    tree = Phylo.read(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/2.For_base_tree_topology/Nymphalidae_fasta_just_topology.nwk", 'newick')

    tree_species_list = [term.name for term in tree.get_terminals()]
    gene_species_list = []
    location = f"/mnt/g/My Drive/Circadian Rhythm Genes Project/{gene_folder}/12.Family_specific_tests_{gene}/Nymphalidae"
    fasta_file = SeqIO.parse(f"{location}/Nymphalidae_{gene}_nostop.fas.aligned.fas.best.fas",  "fasta")

    for records in fasta_file:
        gene_species_list.append(records.id.split("-")[0])
    for species in tree_species_list:
        if species not in gene_species_list:
            tree.prune(species)
    tree_species_list = [term.name for term in tree.get_terminals()]
    fasta_file = SeqIO.parse(f"{location}/Nymphalidae_{gene}_nostop.fas.aligned.fas.best.fas",  "fasta")
    fasta_output = ''
    for records in fasta_file:
        if records.id.split("-")[0] in tree_species_list:
            fasta_output += f">{records.id.split('-')[0]}\n{records.seq}\n"
            # print(records.id.split("-")[0])
    with open(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/3.CodeML_branch_length_optimization/alignments/{gene}_align.fas",'w') as out_file:
        out_file.write(fasta_output)
    Phylo.write(tree, f"/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/3.CodeML_branch_length_optimization/trees/{gene}_tree.nwk", "newick")    
```

```python
from Bio import SeqIO
import os

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/3.hyphy_branch_length_optimization/alignments"
list_of_files = os.listdir(location)
if "desktop.ini" in list_of_files:
    list_of_files.remove("desktop.ini")
species_list = []
for files in list_of_files:
    fasta_file = SeqIO.parse(f"{location}/{files}", 'fasta')

    for record in fasta_file:
        if record.id not in species_list:
            species_list.append(record.id)

print(species_list)
for files in list_of_files:
    fasta_file = SeqIO.parse(f"{location}/{files}", 'fasta')
    species_in_fasta_file = []
    for record in fasta_file:
        species_in_fasta_file.append(record.id)
    for species in species_list:
        if species not in species_in_fasta_file:
            species_list.remove(species)

print(species_list)


for files in list_of_files:
    output = ''
    fasta_file = SeqIO.parse(f"{location}/{files}", 'fasta')
    for record in fasta_file:
        if record.id in species_list:
            output += f">{record.id}\n{record.seq}\n"

    with open(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/4.HYPHY_branch_length_optimization/{files}", 'w') as out_file:
        out_file.write(output)

combined_fasta_file = SeqIO.parse("/mnt/g/My Drive/Circadian Rhythm Genes Project/8.For Phylogeny/2.Tree/combined_aligned.fas",'fasta')
output = ''
for records in combined_fasta_file:
    if records.id in species_list:
        output += f">{records.id}\n{records.seq}\n"
with open(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/19.Evolutionary_rate_co-variation(ERC)/4.HYPHY_branch_length_optimization/for_tree.fas", 'w') as out_file:
    out_file.write(output)        
```
