```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/"
import statistics

gene_omega_dic = {}

with open(f"{location}/35.Busco_all_genes/8.ABSREL_with-ClipKIT/all_omega.tsv", 'r') as omega_file:
    for omega_line in omega_file.readlines():
        omega_split = omega_line.strip().split(",")
        gene_name = omega_split[0]
        omega_list = []
        for omega in omega_split[1:]:
            omega_list.append(float(omega))
        gene_omega_dic[gene_name] = statistics.median(omega_list)
print(len(gene_omega_dic))
```

```python
location = "/mnt/c/Users/Saurav Baral/Desktop/20.RERConverge_trimmed/unTrimmed/clock_trimmed"

import os
import json
import statistics
list_of_files = os.listdir(location)
species_family_dictionary = {}
output = "Gene,Omega"
gene_omega_dic_2 = {}
gene_list = []
for file_name in list_of_files:
    if file_name.endswith(".json"):
        gene = file_name.split(".")[0]
        with open(f'{location}/{file_name}', 'r') as json_file:
            data = json.load(json_file)  # Load JSON data into a Python dictionary or list
        
        for key1,value1 in data["branch attributes"]["0"].items():
            omega = 0.0
            for omega_values in value1["Rate Distributions"]:
                omega += float(omega_values[0])*float(omega_values[1])
            if gene not in gene_omega_dic_2:
                gene_omega_dic_2[gene] = []
            gene_omega_dic_2[gene].append(omega)
```
