```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
data = omega_table.Omega # Simulate right-skewed data
df = pd.DataFrame({'ω (dN/dS)': data, 'Family': ['Nymphalidae'] * len(data)})

# Add text: n and median
median_val = np.median(data)
n_val = len(data)

# Initialize the plot
fig, ax = plt.subplots(figsize=(8, 2.5))

# Half-violin plot (top half) in grayscale
sns.violinplot(data=df, x='ω (dN/dS)', y='Family',
               inner=None, orient='h', cut=0, linewidth=0,
               color='lightgray', width=0.6, ax=ax)

# Overlay boxplot in grayscale
sns.boxplot(data=df, x='ω (dN/dS)', y='Family',
            width=0.05, fliersize=0, color='white',
            boxprops=dict(edgecolor='gray'),
            whiskerprops=dict(color='gray'),
            capprops=dict(color='gray'),
            medianprops=dict(color='gray'),
            orient='h', ax=ax)

# Overlay stripplot (individual data points) in grayscale
sns.stripplot(data=df, x='ω (dN/dS)', y='Family',
              size=2.5, jitter=0.05, color='gray', alpha=0.7,
              orient='h', ax=ax)

marker_positions = [0.28, 0.0, 0.1]  # x-axis (ω values)
marker_colors = ['red', 'blue', 'green']
marker_labels = ['Period', 'CHMP1B', 'PDCD6IP']

# Plot markers and vertical labels
for x, color, label in zip(marker_positions, marker_colors, marker_labels):
    ax.plot(x, 0, marker='o', markersize=6, color=color, markeredgecolor='black', linewidth=0)
    ax.text(x, -0.35, label, fontsize=12, color=color, rotation=90, ha='center', va='bottom')


ax.text(0.7, 0.2, f"n = {n_val} / Median = {median_val:.3f}",
        fontsize=12, va='top', ha='right')

# Aesthetics
ax.set_xlim(0, max(data)+0.1)
ax.set_xlabel('ω (dN/dS)')
ax.set_ylabel('')
ax.set_yticks([])
sns.despine(left=True, bottom=False)
plt.tight_layout()

# Save as PDF
#plt.savefig("nymphalidae_dn_ds_half_violin_grayscale.pdf", format='pdf')
plt.savefig("/mnt/g/My Drive/Circadian Rhythm Genes Project/21.Omega_value_results/dn_ds_violin_greyscale.pdf", format='pdf')
plt.show()
```
