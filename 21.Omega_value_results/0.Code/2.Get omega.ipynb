```python
location = "/mnt/c/Users/Saurav Baral/Desktop/20.RERConverge_trimmed/unTrimmed/clock_trimmed"

import os
import json
import statistics
list_of_files = os.listdir(location)
species_family_dictionary = {}
output = "Gene,Omega"
gene_omega_dic = {}
gene_list = []
for file_name in list_of_files:
    if file_name.endswith(".json"):
        gene = file_name.split(".")[0]
        with open(f'{location}/{file_name}', 'r') as json_file:
            data = json.load(json_file)  # Load JSON data into a Python dictionary or list
        
        for key1,value1 in data["branch attributes"]["0"].items():
            omega = 0.0
            for omega_values in value1["Rate Distributions"]:
                omega += float(omega_values[0])*float(omega_values[1])
            if gene not in gene_omega_dic:
                gene_omega_dic[gene] = []
            gene_omega_dic[gene].append(omega)
for key, value in gene_omega_dic.items():
    gene_list.append([key,statistics.median(value)])
# print(gene_list)
gene_list = (sorted(gene_list, key=lambda x: x[1], reverse=True))
# print(gene_list)
for gene_names in gene_list:
    gene = gene_names[0][0].upper()+gene_names[0][1:]
    for values in gene_omega_dic[gene_names[0]]:
        output += f"\n{gene},{values}"

with open("/mnt/g/My Drive/Circadian Rhythm Genes Project/21.Omega_value_results/3.Clock_omega/omega_all.csv", 'w') as out_file:
    out_file.write(output)
```
