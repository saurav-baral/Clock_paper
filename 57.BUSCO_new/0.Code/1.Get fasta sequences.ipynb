```python
location = "/mnt/griffin/chrwhe/Colias_genomes/Ce_fixed_chromosome_assembly_genomic/run_lepidoptera_odb10/busco_sequences/single_copy_busco_sequences"

import os
import SeqIO

list_of_files= os.listdir(location)

output = ''
for files in list_of_files:
    if files.endswith(".fna"):
        fasta_file = SeqIO.parse(f"{location}/files", 'fasta')
        for records in fasta_file:
            output += f">{records.id}\n{records.seq}\n"

with open ("/mnt/griffin/saubar/10.Busco_run_colias_new/ce_fasta.txt", 'w') as out_file:
    out_file.write(output)
```
