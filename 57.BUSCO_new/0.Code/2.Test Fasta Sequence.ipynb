```python
location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/57.BUSCO_new"

from Bio import SeqIO
import os

eure_dic = {}

fasta_file = SeqIO.parse(f"{location}/ce_fasta.txt", 'fasta')
for records in fasta_file:
    gene_name = (records.id.split("_")[0])
    if records.seq.translate()[-1] == "*":
        eure_dic[gene_name] = records.seq[:-3]
    else:
        eure_dic[gene_name] = records.seq


pala_dic = {}
fasta_file = SeqIO.parse(f"{location}/cp_fasta.txt", 'fasta')
for records in fasta_file:
    gene_name = (records.id.split("_")[0])
    if records.seq.translate()[-1] == "*":
        pala_dic[gene_name] = records.seq[:-3]
    else:
        pala_dic[gene_name] = records.seq

        # pala_dic[gene_name] = records.seq

    # if "*" in records.seq:
    #     print(records.id)
    #     break
```

```python
busco_genes_location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/2.Individual_gene_files_BUSCO_by_3"


list_of_genes = os.listdir(busco_genes_location)
# print(list_of_genes[0])
# assert False
for gene_file_name in list_of_genes:
    if gene_file_name.endswith(".fas"):
        gene_name = gene_file_name.split(".fas")[0]
        if gene_name+".fas" in list_of_genes:
            gene_fasta_file = SeqIO.parse(f"{busco_genes_location}/{gene_name}.fas", 'fasta')
            output = ''
            for species_gene in gene_fasta_file:
                if not(("Colias_nastes" in species_gene.id) or ("Colias_eurytheme" in species_gene.id)):
                    if species_gene.seq.translate()[-1] == "*":
                        output +=  f">{species_gene.id}\n{species_gene.seq[:-3]}\n"
                    else:
                        output +=  f">{species_gene.id}\n{species_gene.seq}\n"
            if gene_name in pala_dic and gene_name in eure_dic:
                output += f">Colias_eurytheme\n{eure_dic[gene_name]}\n>Colias_palaeno\n{pala_dic[gene_name]}\n"
            elif gene_name not in pala_dic and gene_name in eure_dic:
                print(gene_name,"pala_missing")
                output += f">Colias_eurytheme\n{eure_dic[gene_name]}\n"
            elif gene_name not in eure_dic and gene_name in pala_dic:
                print(gene_name,"eure_missing")
                output += f">Colias_palaeno\n{pala_dic[gene_name]}\n"
            else:
                print(gene_name,"both_missing")
            # print(gene_name+".fas")
            # print(output)
            with open(f"/mnt/g/My Drive/Circadian Rhythm Genes Project/22.Busco_related/3.BUSCO_with_colias_fixed/{gene_name}.fas", 'w') as out_file:
                out_file.write(output)
        
            # break
```
