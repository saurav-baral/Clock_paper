```python
location_root = "H:/My Drive/Circadian Rhythm Genes Project/3.Genes from NCBI/1.Period/0.Species/"
species_folder = "1.Bombyx_mori"
location = f"{location_root}/{species_folder}/Transcripts"
species = "Bombyx mori"

import os
from Bio import SeqIO
list_of_species = os.listdir(location_root)

list_of_species.remove('names')
try:
    list_of_species.remove('desktop.ini')
except:
    print(list_of_species)
for species_folder in list_of_species:
    location = f"{location_root}/{species_folder}/Transcripts"
    species = species_folder.split(".")[1].replace("_", " ")
    list_of_transcripts = os.listdir(location)
    max_exons = 0
    longest_transcript = ''
    data_storage_dictionary = {}

    for transcript in list_of_transcripts:
        print(transcript)
        if "desktop" not in transcript:
            data_storage_dictionary[transcript] = {}
    #         print(transcript)
            list_of_exons = os.listdir(f"{location}/{transcript}")

            if len(list_of_exons) > max_exons:
                max_exons = len(list_of_exons)
                longest_transcript = transcript
            for exon_file_name in list_of_exons:
                if "desktop" not in exon_file_name:
    #                 print(exon_file_name)
                    exon = exon_file_name.split(".")[0]
                    exon_data = ''
                    exon_data = SeqIO.parse(f"{location}/{transcript}/{exon_file_name}", 'fasta')
                    for records in exon_data:
    #                     print(records)
                        data_storage_dictionary[transcript][exon] = len(records.seq)

        output = f"Species\tTranscript"
    list_of_exons = []
    for i in range(max_exons-1):
        list_of_exons.append(f"Exon{i+1}")
        output += f"\tExon{i+1}"
    output += f"\n{species}\t{longest_transcript}"
    # print(output)
    for exon in list_of_exons:

        output += f"\t{data_storage_dictionary[longest_transcript][exon]}"

    #     print(key,value)

    list_of_transcripts.remove(longest_transcript)
    try:
        list_of_transcripts.remove("desktop.ini")
    except:
        pass

    for transcript in list_of_transcripts:
#         print(transcript)
    #     print(data_storage_dictionary[transcript])
        output += f"\n{species}\t{transcript}"
        for exon in list_of_exons:
    #         print(key,value)
            try:
                output += f"\t{data_storage_dictionary[transcript][exon]}"
            except:
                break



    #     break


    print(output)
    # print(list_of_exons)

    with open(f"{location_root}/{species_folder}/extended_exon_set.tsv", 'w') as out_file:
        out_file.write(output)
```
