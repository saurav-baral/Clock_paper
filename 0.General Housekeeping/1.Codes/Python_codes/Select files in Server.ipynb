```python
import os
import subprocess

gene_folder = "9.CLOCK Exon Analysis"

location = "/mnt/g/My Drive/Circadian Rhythm Genes Project/"
out_location = "/mnt/g/My Drive/temp"


folder_root = (os.listdir(f"{location}/{gene_folder}"))
if "desktop.ini" in folder_root:
    folder_root.remove("desktop.ini")

for folder_name_root in folder_root:
    print(folder_name_root)
    if os.path.isdir(f"{location}/{gene_folder}/{folder_name_root}"):
        list_of_folders_1 = ((os.listdir(f"{location}/{gene_folder}/{folder_name_root}")))
        if "desktop.ini" in list_of_folders_1:
            list_of_folders_1.remove("desktop.ini")
        if "1.Blast_result" not in list_of_folders_1:
            print(folder_name_root)
            subprocess.run(f"cp -r '{location}/{gene_folder}/{folder_name_root}' '{out_location}/{gene_folder}'", shell = True)
        else:
            try:
                os.mkdir(f'{out_location}/{gene_folder}/{folder_name_root}')
            except:
                pass
            for folder_name_1 in list_of_folders_1:
                try:
                    os.mkdir(f'{out_location}/{gene_folder}/{folder_name_root}/1.Blast_result')
                except:
                    pass
                if "1.Blast_result" not in folder_name_1:
                    print(folder_name_1)
                    subprocess.run(f"cp -r '{location}/{gene_folder}/{folder_name_root}/{folder_name_1}' '{out_location}/{gene_folder}/{folder_name_root}'", shell = True)
                list_of_folders_2 = ((os.listdir(f"{location}/{gene_folder}/{folder_name_root}/1.Blast_result")))
                if "desktop.ini" in list_of_folders_2:
                    list_of_folders_2.remove("desktop.ini")    
                
                for folder_name_2 in list_of_folders_2:
                    try:
                        os.mkdir(f'{out_location}/{gene_folder}/{folder_name_root}/1.Blast_result/{folder_name_2}')
                    except:
                        pass
                    subprocess.run(f"cp -r '{location}/{gene_folder}/{folder_name_root}/1.Blast_result/{folder_name_2}/final_coordinates.csv' '{out_location}/{gene_folder}/{folder_name_root}/1.Blast_result/{folder_name_2}/'", shell = True)
    else:
        subprocess.run(f"cp -r '{location}/{gene_folder}/{folder_name_root}' '{out_location}/{gene_folder}'", shell = True)
            

```
