```python
from Bio import SeqIO

folder = "12.Yellow_Un3"

alignment =  SeqIO.parse(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/MRJP_domain_aligned_translated.fas", 'fasta')

for records in alignment:
    if "Bombyx_mori" in records.id:
        bombyx_sequence = str(records.seq)

#         print(records.id)

with open(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/LEISR_out.txt",'r') as leisr_file:
    leisr_data_list = leisr_file.readlines()

out_sequence = f'>bombyx_mori_{folder.split(".")[1]}_MRJP_domain\n'
out_leisr = 'Site\tEvolutionary Rate\n'
counter = 1
for i in range(len(bombyx_sequence)):
    if bombyx_sequence[i] != "-":
        out_sequence += bombyx_sequence[i]
#         print(out_sequence)
        leisr_value = leisr_data_list[i+1].split("\t")[1].rstrip()
        out_leisr += f"{counter}\t{leisr_value}\n"
        counter += 1

with open(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/LEISR_out_bombyx.txt",'w') as leisr_out_file:
    leisr_out_file.write(out_leisr)

with open(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/bombyx_mrjp.fas",'w') as sequence_out_file:
    sequence_out_file.write(out_sequence)

print(len(leisr_data_list), len(bombyx_sequence))
```
