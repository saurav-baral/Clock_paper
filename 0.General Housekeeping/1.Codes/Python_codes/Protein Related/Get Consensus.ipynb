```python
from __future__ import annotations
from pathlib import Path
from itertools import chain

from Bio import SeqIO
from itertools import chain
from Bio import AlignIO
from Bio.Align import MultipleSeqAlignment
from Bio.Align.AlignInfo import SummaryInfo



import Bio
from Bio import AlignIO
from Bio.Align import MultipleSeqAlignment
from Bio.Align.AlignInfo import SummaryInfo

folder = "4.Yellow_d"
gene_name = "Yellow_d"

SeqRecord = Bio.SeqRecord.SeqRecord

common_alignment = MultipleSeqAlignment(
        chain(*AlignIO.parse(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/MRJP_domain_aligned_translated.fas", "fasta"))
    )
summary = SummaryInfo(common_alignment)
consensus = summary.dumb_consensus(0.0000000001, "X")
print(consensus)

alignment_file = f"K:/My Drive/MRJP/6.Rate4Site/{folder}/MRJP_domain_aligned_translated.fas"
alignment = SeqIO.parse(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/MRJP_domain_aligned_translated.fas", 'fasta')

common_alignment = MultipleSeqAlignment(
        chain(*AlignIO.parse(alignment_file, "fasta"))
    )

out = f'>consensus_{gene_name}\n'
for i in range(len(consensus)):
    if consensus[i] == "X":
        for records in alignment:
            sequence = records.seq
            if sequence[i] != "-":
#                 print(i,consensus[i], sequence[i])
                out += sequence[i]
                break
    else:
        out += consensus[i]

    
with open(f"K:/My Drive/MRJP/6.Rate4Site/{folder}/MRJP_domain_translated_consensus.fas",'w') as out_file:
    out_file.write(out)
print(out)
```
