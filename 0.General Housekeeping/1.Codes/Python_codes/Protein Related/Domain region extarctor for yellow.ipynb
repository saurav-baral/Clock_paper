```python
import os
from Bio import SeqIO

gene_folder = "12.Yellow_Un3"
drive= "K"
domain_start = 442
domain_end = 1305

list_of_files = os.listdir(f"{drive}:/My Drive/MRJP/Genes/9.AnalyzeCodonData/{gene_folder}")

alignment_file = ''

for file in list_of_files:
    
    if "aligned_untranslated.fa" in file:
        alignment_file = file

alignment = SeqIO.parse(f"{drive}:/My Drive/MRJP/Genes/9.AnalyzeCodonData/{gene_folder}/{alignment_file}", 'fasta')

out = ''
for records in alignment:
    out+= ">"+records.id+"\n"
    out+= records.seq[domain_start-1:domain_end]+"\n"
#     print(records.seq[domain_start-1:domain_end])
#     break
print(out)
with open(f"{drive}:/My Drive/MRJP/6.Rate4Site/{gene_folder}/MRJP_domain_aligned.fas", 'w') as out_file:
    out_file.write(str(out))
```
