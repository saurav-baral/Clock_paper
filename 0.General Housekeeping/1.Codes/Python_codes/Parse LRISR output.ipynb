```python
import json

folder = "12.Yellow_Un3"
with open (f"K:/My Drive/MRJP/6.Rate4Site/{folder}/MRJP_domain_aligned_translated.fas.LEISR.json", 'r') as json_file:
    file_read = json.load(json_file)


# file_read = json.loads(json_file_as_list)
output = "Site\tEvolutionary Rate\n"
for i in range(len(file_read["MLE"]["content"]["0"])):
    rate = float(file_read["MLE"]["content"]["0"][i][0]) 
    if rate < 10:
        output += f"{i+1}\t{rate}\n"
    else:
        output += f"{i+1}\t0\n"
        print(f"{i+1}\t{rate}")

with open (f"K:/My Drive/MRJP/6.Rate4Site/{folder}/LEISR_out.txt", 'w') as out_file:
    out_file.write(output)
```
