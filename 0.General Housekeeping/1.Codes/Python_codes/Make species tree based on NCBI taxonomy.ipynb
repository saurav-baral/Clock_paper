```python
import re
import sys
import os
from Bio import Entrez
Entrez.email = 'saubaralwork@gmail.com'


def get_ncbi_tax(taxon):
    '''Getn NCBI taxonomy'''
    # code from https://harryincupboard.blog/3
    # If the input is a string
    if not re.match(r'\d+', taxon):
        # Get taxonomy ID using Entrez
        taxon2 = '"' + taxon + '"'
        handle = Entrez.esearch(
            db='taxonomy', term=taxon2, rettype='gb', retmode='text')
        record = Entrez.read(handle, validate=False)
        handle.close()
        # If there's no result
        if not record['IdList']:
            sys.exit(
                '[ERROR] The taxon "{}" you provided is invalid. '
                'Please check NCBI Taxonomy'.format(taxon))
        tax_id = record['IdList']
    else:
        tax_id = taxon

    # Now connect NCBI again using the tax_id
    # Entrez.efetch will give you various information
    handle2 = Entrez.efetch(db='taxonomy', id=tax_id, retmode='xml')
    record2 = Entrez.read(handle2, validate=False)
    handle2.close()

    tax_list = record2[0]['LineageEx']
    return (tax_list)

# taxonomic_list = get_ncbi_tax("Brenthis daphne")
# print(taxonomic_list)
list_of_genomes = os.listdir("F:/Genomes_2023-12-26")
# list_of_genomes = ["Blastobasis adustella"]

output = 'Order\tSuborder\tInfraorder\tparvorder\tSuperfamily\tFamily\tSubfamily\tGenus\n'
for genomes in list_of_genomes:
    species = genomes.replace("_"," ")
    print(species)
    order = f"Missing_{species}"
    suborder = f"Missing_{species}"
    infraorder = f"Missing_{species}"
    parvorder = f"Missing_{species}"
    super_family = f"Missing_{species}"
    family = f"Missing_{species}"
    subfamily = f"Missing_{species}"
    genus = f"Missing_{species}"
    
    order_done = 0
    suborder_done = 0
    infraorder_done = 0
    parvorder_done = 0
    super_family_done = 0
    family_done = 0
    subfamily_done = 0
    genus_done = 0
    
    
    species_done = 0
    while species_done == 0:
        print(species)
        try:
            taxonomic_list = get_ncbi_tax(species)
            for taxonomic_unit in taxonomic_list:
    #             print(taxonomic_unit["Rank"],(taxonomic_unit["ScientificName"]))
                if taxonomic_unit["Rank"] == "order":
                    order = taxonomic_unit["ScientificName"]
                    order_done = 1
                if taxonomic_unit["Rank"] == "suborder":
                    suborder = taxonomic_unit["ScientificName"]
                    suborder_done = 1
                if taxonomic_unit["Rank"] == "infraorder":
                    infraorder = taxonomic_unit["ScientificName"]
                    infraorder_done = 1
                if taxonomic_unit["Rank"] == "parvorder":
                    parvorder = taxonomic_unit["ScientificName"]
                    parvorder_done = 1
                if taxonomic_unit["Rank"] == "superfamily":
                    super_family = taxonomic_unit["ScientificName"]
                    super_family_done = 1
                if taxonomic_unit["Rank"] == "family":
                    family = taxonomic_unit["ScientificName"]
                    family_done = 1
                if taxonomic_unit["Rank"] == "subfamily":
                    subfamily = taxonomic_unit["ScientificName"]
                    subfamily_done = 1
                if taxonomic_unit["Rank"] == "genus":
                    genus = taxonomic_unit["ScientificName"]
                    genus_done = 1
                print(subfamily,subfamily_done)
                if genus_done == 0:
                    genus = f"Missing_{species}"
                
                if subfamily_done == 0:
                    subfamily = f"Missing_{genus}"
                
                if family_done == 0:
                    family = f"Missing_{subfamily}"
                
                if super_family_done == 0:
                    super_family = f"Missing_{family}"
                
                print(subfamily,subfamily_done)
                
                species_done = 1
                
        except:
            pass
#             order = "Missing"
#             suborder = "Missing"
#             infraorder = "Missing"
#             parvorder = "Missing"
#             super_family = "Missing"
#             family = "Missing"
#             subfamily = "Missing"
#             genus = "Missing"
    
    
    output += f"{order}\t{suborder}\t{infraorder}\t{parvorder}\t{super_family}\t{family}\t{subfamily}\t{genus}\t{species}\n"
    print(f"{order}\t{suborder}\t{infraorder}\t{parvorder}\t{super_family}\t{family}\t{subfamily}\t{genus}\t{species}\n")
#     print(f"{genomes}, {family}, {super_family}")
#     print(output)
#     print(taxonomic_list)
#     break 

with open("H:/My Drive/Circadian Rhythm Genes Project/5.Blast output for all/taxonomic_list.csv",'w') as out_file:
    out_file.write(output)

#     
#     break
```
