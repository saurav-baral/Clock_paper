```python
import pandas as pd
from statsmodels.stats.multitest import multipletests
import numpy as np
import qvalue

# Step 1: Create or read your DataFrame
# Example: Read from CSV
df = pd.read_csv("/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/trees_split/correlation_2Apr2025_sp_rem_modi2.csv")

# Or manually create a sample DataFrame

# Step 2: Apply Holmâ€“Bonferroni correction using statsmodels
rejected, pvals_corrected, _, _ = multipletests(df['P'], alpha=0.05, method='fdr_bh')

# Step 3: Add results to DataFrame
# qobj = qvalue(df['P'].tolist())
df['Holm_corrected'] = pvals_corrected
df['Significant'] = rejected

# Display results
print(df)
# df.to_csv("/mnt/c/Users/sauba/Desktop/20.RERConverge_trimmed/3.New_colias/trees_split/correlation_2Apr2025_sp_rem_modi2_added_correction.csv")
```

```python
from multipy.fdr import lsu
from multipy.data import neuhaus

pvals = neuhaus()
print(pvals)
significant_pvals = lsu(pvals)
print(significant_pvals)
print(zip(['{:.4f}'.format(p) for p in pvals], significant_pvals))
for p, sig in zip(pvals,significant_pvals):
    print(p,sig)

```
