```python
list_of_genes_new = os.listdir(f"{location}/3.New_colias/4.BUSCO_with_colias_fixed_aligned_trimmed")
print(len(list_of_genes_new))
gene_list_new = []
# counter += 0
for gene_name in list_of_genes_new:
    gene_list_new.append(gene_name.replace(".fas.best.fas.ClipKIT.fas", '')) 
missing = []
for gene_name in gene_list:
    if gene_name not in gene_list_new:
        print(gene_name)
        missing.append(gene_name)
```

```python
from Bio import SeqIO

for gene_name in missing:
    fasta_file = SeqIO.parse(f"{location}/unTrimmed/2.Trimmed_alignment/{gene_name}_nostop.fas.aligned.fas.best.fas.ClipKIT.fas", 'fasta')
    output = ''
    for records in fasta_file:
        if records.id != "Colias_nastes":
            output += f">{records.id}\n{records.seq}\n"
    with open(f"{location}/3.New_colias/4.BUSCO_with_colias_fixed_aligned_trimmed_missing/{gene_name}.fas.best.fas.ClipKIT.fas", 'w') as out_file:
        out_file.write(output)
```
