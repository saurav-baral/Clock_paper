```python
location = "/mnt/griffin/saubar/busco_runs_CDS_files"

import os
from Bio import SeqIO

list_of_species = os.listdir(location)
list_of_species.remove("count_genes.py")
# print(list_of_species)
busco_count_dictionary = {}
busco_per_gene_count = {}
busco_count_dictionary_checked = {}
for i,genome_folder in enumerate(list_of_species):
    print(i, len(list_of_species))
    if genome_folder.endswith("_genomic.fna"):
        list_of_busco_genes = os.listdir(f'{location}/{genome_folder}/CDS')
        genome_count = 0
        for gene_name in list_of_busco_genes:
            fasta_sequence = SeqIO.parse(f"{location}/{genome_folder}/CDS/{gene_name}", 'fasta')
            for records in fasta_sequence:
                if "*" not in records.seq.translate()[:-1]:
                    genome_count += 1
            if gene_name not in busco_per_gene_count:
                busco_per_gene_count[gene_name] = 0
            busco_per_gene_count[gene_name] += 1
        busco_count_dictionary[genome_folder] = len(list_of_busco_genes)
        busco_count_dictionary_checked[genome_folder] = genome_count
output1 = "Genome,BUSCO_gene"

for key, value in busco_count_dictionary.items():
    output1 += f"\n{key},{value}"
with open("/mnt/griffin/saubar/busco_runs_CDS_files/per_genome_count_without_checking.txt", 'w') as out_file:
    out_file.write(output1)

output1 = "Genome,BUSCO_gene"

for key, value in busco_count_dictionary_checked.items():
    output1 += f"\n{key},{value}"
with open("/mnt/griffin/saubar/busco_runs_CDS_files/per_genome_count_with_checking.txt", 'w') as out_file:
    out_file.write(output1)


output1 = "Gene,Total_count"

for key, value in busco_per_gene_count.items():
    output1 += f"\n{key},{value}"
with open("/mnt/griffin/saubar/busco_runs_CDS_files/gene_count_without_checking.txt", 'w') as out_file:
    out_file.write(output1)

```
